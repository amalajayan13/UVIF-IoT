# UVIF-IoT: RAM-Safe Full Notebook

**UVIF-Edge / UVIF-IoT: A Variational Intelligence Framework for Uncertainty-Aware and Risk-Sensitive Anomaly Detection at the IoT Edge**

This notebook is designed to avoid Colab crashes by using memory-bounded data loading, capped adversarial evaluation, safe provenance checks, robust state persistence, and guarded output generation. It supports real N-BaIoT and UNSW-NB15 datasets when present in Google Drive, and creates explicitly non-reportable synthetic smoke-test data only when files are missing.


```python
# ── Cell 1 — Environment, paths, configuration, and reproducibility ──────────
print("Cell 1: setup entered")

import os, gc, json, time, random, warnings
from pathlib import Path
import numpy as np
import pandas as pd
warnings.filterwarnings("ignore")

SEED = 42
random.seed(SEED); np.random.seed(SEED)

try:
    import torch, torch.nn as nn, torch.nn.functional as F
    from torch.utils.data import TensorDataset, DataLoader
    torch.manual_seed(SEED)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(SEED)
    DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
except Exception as e:
    raise RuntimeError(f"PyTorch is required: {e}")

# ── GPU guard: refuse to run silently on CPU ──────────────────────────────────
if DEVICE.type == "cpu":
    raise RuntimeError(
        "No GPU detected. Go to Runtime → Change runtime type → T4 GPU "
        "and reconnect before running this notebook."
    )

try:
    from google.colab import drive
    IN_COLAB = True
except Exception:
    IN_COLAB = False

USE_GOOGLE_DRIVE = True
PROJECT_NAME = "UVIF_IoT"

if IN_COLAB and USE_GOOGLE_DRIVE:
    try:
        drive.mount('/content/drive', force_remount=False)
        BASE_DIR = Path('/content/drive/MyDrive/Outputs') / PROJECT_NAME
    except Exception as e:
        print(f"Google Drive mount failed: {e}")
        BASE_DIR = Path('/content') / PROJECT_NAME
else:
    BASE_DIR = Path('/content') / PROJECT_NAME if Path('/content').exists() else Path.cwd() / PROJECT_NAME

FIG_DIR   = BASE_DIR / 'figures'
TABLE_DIR = BASE_DIR / 'tables'
OUTPUT_DIR= BASE_DIR / 'outputs'
CKPT_DIR  = BASE_DIR / 'checkpoints'
for d in [BASE_DIR, FIG_DIR, TABLE_DIR, OUTPUT_DIR, CKPT_DIR]:
    d.mkdir(parents=True, exist_ok=True)

RUN_LOG_PATH = OUTPUT_DIR / 'run_log.txt'
SUMMARY_PATH = OUTPUT_DIR / 'outputs_summary.txt'
Q1_PATH      = OUTPUT_DIR / 'q1_readiness_diagnosis.txt'
RUN_LOG_PATH.write_text('', encoding='utf-8')

def now(): return time.strftime('%H:%M:%S')

def log(msg):
    line = f"[{now()}] {msg}"
    print(line)
    with open(RUN_LOG_PATH, 'a', encoding='utf-8') as f:
        f.write(line + '\n')

def summarise(msg=''):
    with open(SUMMARY_PATH, 'a', encoding='utf-8') as f:
        f.write(str(msg) + '\n')

CFG = {
    # RAM safety
    'max_total_rows_nbaiot':     300_000,
    'max_rows_per_nbaiot_file':   25_000,
    'max_total_rows_unsw_train': 175_000,
    'max_total_rows_unsw_test':   80_000,
    'use_smote': False,
    # Dataloaders / training
    'batch_size': 512, 'num_workers': 0, 'pin_memory': False,
    'epochs': 30, 'lr': 1e-3, 'weight_decay': 1e-5,
    'p_drop': 0.30, 'use_class_weights': True, 'early_stop_patience': 8,
    # UVIF Lagrangian (§III-B)
    'uvif_warmup_epochs': 8,
    'tau_start': 1.0, 'tau_end': 0.05,
    'beta_evid': 0.10, 'beta_cal': 0.20,
    'rho_wass': 0.01,
    'lambda_dro': 0.05, 'lambda_cal': 0.03, 'lambda_e': 0.01,
    'eta_mu': 0.03, 'mu_max': 2.0,
    'epsilon_r': 0.20, 'epsilon_c': 0.15, 'epsilon_e': 0.35,
    'mu_warmup_epochs': 8,
    # Evaluation
    'mc_samples': 5, 'ece_bins': 15,
    'pgd_steps': 10, 'pgd_alpha': 0.004,
    # 50 batches × 512 = ~25k samples: statistically valid for UNSW rare classes
    'adv_batch_cap': 50,
    'rf_n_estimators': 80, 'skip_rf_for_speed': False,
    'run_pareto': True,
}

DATA_CANDIDATE_PATHS = [
    Path('/content/drive/MyDrive/Datasets'),
    Path('/content/drive/MyDrive/Datasets/Telecom'),
    Path('/content/drive/MyDrive/Datasets/N_BaIoT'),
    Path('/content/drive/MyDrive/Datasets/NBaIoT'),
    Path('/content/drive/MyDrive/Datasets/UNSW_NB15'),
    Path('/content/drive/MyDrive/Datasets/Telecom/UNSW_NB15'),
    Path('/mnt/data'),
]

log(f"Output root: {BASE_DIR}")
log(f"Using device: {DEVICE} ({torch.cuda.get_device_name(0)})")
log("Dataset path candidates configured.")
log(f"N-BaIoT RAM cap: total={CFG['max_total_rows_nbaiot']:,}, per_file={CFG['max_rows_per_nbaiot_file']:,}")
log(f"UNSW RAM cap: train={CFG['max_total_rows_unsw_train']:,}, test={CFG['max_total_rows_unsw_test']:,}")
print("Cell 1 is completed")
```

## 2. RAM-safe data loading

The loaders are intentionally capped. This is not a weakness: it prevents Colab memory crashes and keeps experiments reproducible. Synthetic fallback data are only for smoke testing and automatically marked non-reportable.


```python
# ── Cell 2.1 — N-BaIoT loader: RAM-safe, balanced, provenance-aware ──────────
print("Cell 2.1: N-BaIoT loader entered")

from sklearn.datasets import make_classification

NBAIOT_SYNTHETIC = False

def label_from_nbaiot_path(path):
    s = str(path).lower()
    if 'benign_traffic' in s: return 0
    if 'mirai' in s:          return 1
    if 'gafgyt' in s:         return 2
    return None

def find_nbaiot_files(root_paths):
    files = []
    for root in root_paths:
        root = Path(root)
        if not root.exists(): continue
        for fp in root.rglob('*.csv'):
            lab = label_from_nbaiot_path(fp)
            if lab is not None: files.append((fp, lab))
    return sorted(set(files), key=lambda x: (str(x[0]).lower(), x[1]))

def safe_read_csv_sample(path, max_rows, random_state=SEED):
    df = pd.read_csv(path, low_memory=False)
    if len(df) > max_rows:
        df = df.sample(n=max_rows, random_state=random_state).reset_index(drop=True)
    return df

nb_files = find_nbaiot_files(DATA_CANDIDATE_PATHS)
log(f"N-BaIoT candidate CSV files found: {len(nb_files)}")

frames = []; class_counts = {0: 0, 1: 0, 2: 0}

if len(nb_files) == 0:
    NBAIOT_SYNTHETIC = True
    log('N-BaIoT files not found. Generating synthetic smoke-test data.')
    X_s, y_s = make_classification(n_samples=12_000, n_features=115, n_informative=40,
        n_redundant=20, n_classes=3, weights=[1/3,1/3,1/3], random_state=SEED)
    nbaiot_raw = pd.DataFrame(X_s, columns=[f'f{i:03d}' for i in range(115)])
    nbaiot_raw['label'] = y_s; nbaiot_raw['device'] = 'synthetic'
else:
    per_class_cap = CFG['max_total_rows_nbaiot'] // 3
    for fp, lab in nb_files:
        if sum(class_counts.values()) >= CFG['max_total_rows_nbaiot']: break
        if class_counts[lab] >= per_class_cap: continue
        remaining_for_class = per_class_cap - class_counts[lab]
        remaining_total = CFG['max_total_rows_nbaiot'] - sum(class_counts.values())
        take = min(CFG['max_rows_per_nbaiot_file'], remaining_for_class, remaining_total)
        if take <= 0: continue
        try:
            df = safe_read_csv_sample(fp, take)
            if len(df) > take: df = df.sample(n=take, random_state=SEED).reset_index(drop=True)
            df['label'] = lab; df['source_file'] = str(fp)
            frames.append(df); class_counts[lab] += len(df)
            log(f"  {fp.name} -> label={lab} ({len(df):,} rows); total={sum(class_counts.values()):,}; classes={class_counts}")
        except Exception as e:
            log(f"  Skipped {fp}: {e}")
        gc.collect()
    if frames:
        nbaiot_raw = pd.concat(frames, ignore_index=True)
        min_count = nbaiot_raw['label'].value_counts().min()
        nbaiot_raw = (nbaiot_raw.groupby('label', group_keys=False)
                      .apply(lambda x: x.sample(n=min(len(x), min_count), random_state=SEED))
                      .reset_index(drop=True))
        log(f"N-BaIoT RAM-safe final shape: {nbaiot_raw.shape[0]:,} rows, {nbaiot_raw.shape[1]} columns")
        log(f"N-BaIoT label distribution: {nbaiot_raw['label'].value_counts().sort_index().to_dict()}")
    else:
        NBAIOT_SYNTHETIC = True
        log('No readable N-BaIoT files. Generating synthetic smoke-test data.')
        X_s, y_s = make_classification(n_samples=12_000, n_features=115, n_classes=3, random_state=SEED)
        nbaiot_raw = pd.DataFrame(X_s, columns=[f'f{i:03d}' for i in range(115)])
        nbaiot_raw['label'] = y_s; nbaiot_raw['device'] = 'synthetic'

gc.collect()
print("Cell 2.1 is completed")
```

```python
# ── Cell 2.2 — UNSW-NB15 loader: RAM-safe and provenance-aware ───────────────
print("Cell 2.2: UNSW-NB15 loader entered")

UNSW_SYNTHETIC = False

def find_unsw(root_paths):
    candidates = [
        ('UNSW_NB15_training-set.csv', 'UNSW_NB15_testing-set.csv'),
        ('UNSW-NB15_training-set.csv', 'UNSW-NB15_testing-set.csv'),
        ('UNSW_NB15_training.csv', 'UNSW_NB15_testing.csv'),
    ]
    for root in root_paths:
        root = Path(root)
        if not root.exists(): continue
        for tr_name, te_name in candidates:
            tr, te = root / tr_name, root / te_name
            if tr.exists() and te.exists(): return tr, te
            tr_hits = list(root.rglob(tr_name)); te_hits = list(root.rglob(te_name))
            if tr_hits and te_hits: return tr_hits[0], te_hits[0]
    return None, None

def read_unsw_csv(path, max_rows, name):
    df = pd.read_csv(path, low_memory=False)
    if len(df) > max_rows:
        if 'attack_cat' in df.columns:
            frac = max_rows / len(df)
            parts = []
            for _, group in df.groupby('attack_cat'):
                n = min(len(group), max(1, int(round(len(group) * frac))))
                parts.append(group.sample(n=n, random_state=SEED))
            df = pd.concat(parts, ignore_index=True)
            if len(df) > max_rows: df = df.sample(n=max_rows, random_state=SEED).reset_index(drop=True)
        else:
            df = df.sample(n=max_rows, random_state=SEED).reset_index(drop=True)
    log(f'{name}: {df.shape}')
    return df

unsw_train_path, unsw_test_path = find_unsw(DATA_CANDIDATE_PATHS)

if unsw_train_path is not None and unsw_test_path is not None:
    UNSW_SYNTHETIC = False
    log(f'UNSW-NB15 train path: {unsw_train_path}')
    log(f'UNSW-NB15 test path : {unsw_test_path}')
    unsw_train_raw = read_unsw_csv(unsw_train_path, CFG['max_total_rows_unsw_train'], 'UNSW-NB15 train RAM-safe')
    unsw_test_raw  = read_unsw_csv(unsw_test_path,  CFG['max_total_rows_unsw_test'],  'UNSW-NB15 test RAM-safe')
    if 'attack_cat' in unsw_train_raw.columns:
        log(f'UNSW train class distribution: {unsw_train_raw["attack_cat"].value_counts().to_dict()}')
else:
    UNSW_SYNTHETIC = True
    log('UNSW-NB15 files not found. Generating synthetic smoke-test data.')
    X_s, y_s = make_classification(n_samples=10_000, n_features=45, n_informative=20,
        n_redundant=10, n_classes=2, weights=[0.80,0.20], random_state=SEED)
    unsw_train_raw = pd.DataFrame(X_s[:8000], columns=[f'f{i:02d}' for i in range(45)])
    unsw_train_raw['attack_cat'] = np.where(y_s[:8000]==1,'DoS','Normal')
    unsw_test_raw  = pd.DataFrame(X_s[8000:],  columns=[f'f{i:02d}' for i in range(45)])
    unsw_test_raw['attack_cat']  = np.where(y_s[8000:]==1,'DoS','Normal')

gc.collect()
print("Cell 2.2 is completed")
```

```python
# ── Cell 2.3 — Data provenance guard ─────────────────────────────────────────
print("Cell 2.3: provenance guard entered")

DATA_PROVENANCE = {
    'nbaiot_available':     'nbaiot_raw' in globals(),
    'unsw_available':       'unsw_train_raw' in globals() and 'unsw_test_raw' in globals(),
    'nbaiot_is_synthetic':  bool(globals().get('NBAIOT_SYNTHETIC', True)),
    'unsw_is_synthetic':    bool(globals().get('UNSW_SYNTHETIC', True)),
}
DATA_PROVENANCE['any_synthetic'] = DATA_PROVENANCE['nbaiot_is_synthetic'] or DATA_PROVENANCE['unsw_is_synthetic']
REPORTABLE_RESULTS = not DATA_PROVENANCE['any_synthetic']

if REPORTABLE_RESULTS:
    log('DATA STATUS: Real datasets detected. Results may be used after final full-run verification.')
else:
    log('DATA STATUS: Synthetic smoke-test mode detected. Results are NON-REPORTABLE.')

print(json.dumps(DATA_PROVENANCE, indent=2))
print("Cell 2.3 is completed")
```

```python
# ── Cell 3 — Preprocessing, splits, and DataLoaders ──────────────────────────
print("Cell 3: preprocessing entered")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

nb_df = nbaiot_raw.copy()
for col in ['source_file','device']:
    if col in nb_df.columns: nb_df = nb_df.drop(columns=[col])
nb_df = nb_df.replace([np.inf,-np.inf], np.nan).dropna(axis=1, how='all')
nb_df = nb_df.fillna(nb_df.median(numeric_only=True))

nb_label = nb_df['label'].astype(int).values
nb_X_df  = nb_df.drop(columns=['label']).select_dtypes(include=[np.number]).copy()
X_nb = nb_X_df.values.astype(np.float32); y_nb = nb_label.astype(int)
nb_feature_cols = list(nb_X_df.columns)

X_nb_train, X_nb_tmp, y_nb_train, y_nb_tmp = train_test_split(X_nb, y_nb, test_size=0.30, stratify=y_nb, random_state=SEED)
X_nb_val,   X_nb_test, y_nb_val,  y_nb_test = train_test_split(X_nb_tmp, y_nb_tmp, test_size=2/3, stratify=y_nb_tmp, random_state=SEED)
nb_scaler = StandardScaler()
X_nb_train = nb_scaler.fit_transform(X_nb_train).astype(np.float32)
X_nb_val   = nb_scaler.transform(X_nb_val).astype(np.float32)
X_nb_test  = nb_scaler.transform(X_nb_test).astype(np.float32)
log(f"N-BaIoT — train: {X_nb_train.shape} | val: {X_nb_val.shape} | test: {X_nb_test.shape}")
log(f"N-BaIoT training classes: {dict(zip(*np.unique(y_nb_train, return_counts=True)))}")

unsw_train = unsw_train_raw.copy(); unsw_test = unsw_test_raw.copy()
target_col = 'attack_cat' if 'attack_cat' in unsw_train.columns else ('label' if 'label' in unsw_train.columns else unsw_train.columns[-1])
for df in [unsw_train, unsw_test]:
    for c in ['id','srcip','sport','dstip','dsport','Stime','Ltime']:
        if c in df.columns: df.drop(columns=[c], inplace=True)
if target_col not in unsw_test.columns: unsw_test[target_col] = 'Normal'
all_unsw = pd.concat([unsw_train, unsw_test], axis=0, ignore_index=True).replace([np.inf,-np.inf], np.nan)
y_all_raw = all_unsw[target_col].astype(str).fillna('Unknown')
X_all_df  = all_unsw.drop(columns=[target_col])
cat_cols = X_all_df.select_dtypes(include=['object','category']).columns.tolist()
num_cols = [c for c in X_all_df.columns if c not in cat_cols]
for c in num_cols: X_all_df[c] = pd.to_numeric(X_all_df[c], errors='coerce')
X_all_df[num_cols] = X_all_df[num_cols].fillna(X_all_df[num_cols].median(numeric_only=True))
if cat_cols:
    X_all_df[cat_cols] = X_all_df[cat_cols].astype(str).fillna('Unknown')
    X_all_df = pd.get_dummies(X_all_df, columns=cat_cols, dummy_na=False)
le_unsw = LabelEncoder(); y_all = le_unsw.fit_transform(y_all_raw)
n_train_raw = len(unsw_train)
X_unsw_train_full = X_all_df.iloc[:n_train_raw].values.astype(np.float32); y_unsw_train_full = y_all[:n_train_raw].astype(int)
X_unsw_test       = X_all_df.iloc[n_train_raw:].values.astype(np.float32); y_unsw_test       = y_all[n_train_raw:].astype(int)
unsw_feature_cols = list(X_all_df.columns)
try:
    X_unsw_train, X_unsw_val, y_unsw_train, y_unsw_val = train_test_split(
        X_unsw_train_full, y_unsw_train_full, test_size=0.10, stratify=y_unsw_train_full, random_state=SEED)
except Exception:
    X_unsw_train, X_unsw_val, y_unsw_train, y_unsw_val = train_test_split(
        X_unsw_train_full, y_unsw_train_full, test_size=0.10, random_state=SEED)
unsw_scaler = StandardScaler()
X_unsw_train = unsw_scaler.fit_transform(X_unsw_train).astype(np.float32)
X_unsw_val   = unsw_scaler.transform(X_unsw_val).astype(np.float32)
X_unsw_test  = unsw_scaler.transform(X_unsw_test).astype(np.float32)
log(f"UNSW-NB15 — train: {X_unsw_train.shape} | val: {X_unsw_val.shape} | test: {X_unsw_test.shape}")
log(f"Classes: {list(le_unsw.classes_)}")

def make_loader(X, y, shuffle=False):
    ds = TensorDataset(torch.tensor(X, dtype=torch.float32), torch.tensor(y, dtype=torch.long))
    return DataLoader(ds, batch_size=CFG['batch_size'], shuffle=shuffle,
                      num_workers=CFG['num_workers'], pin_memory=CFG['pin_memory'])

nb_loaders   = (make_loader(X_nb_train, y_nb_train, True),   make_loader(X_nb_val, y_nb_val),   make_loader(X_nb_test, y_nb_test))
unsw_loaders = (make_loader(X_unsw_train, y_unsw_train, True), make_loader(X_unsw_val, y_unsw_val), make_loader(X_unsw_test, y_unsw_test))

log('SMOTE disabled. Class imbalance will be handled with class-weighted losses.')
log('DataLoaders ready.')
gc.collect()
print("Cell 3 is completed")
```

```python
# ── Cell 4 — Attack helpers and baseline MLP ─────────────────────────────────
print("Cell 4: helpers and baseline entered")

from sklearn.metrics import (f1_score, accuracy_score, balanced_accuracy_score,
                              precision_score, recall_score, roc_auc_score, average_precision_score)
from sklearn.calibration import calibration_curve
from sklearn.utils.class_weight import compute_class_weight
from sklearn.ensemble import RandomForestClassifier, IsolationForest

class MLP(nn.Module):
    def __init__(self, in_dim, n_classes, p_drop=0.0):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim,256), nn.ReLU(), nn.Dropout(p_drop),
            nn.Linear(256,128),    nn.ReLU(), nn.Dropout(p_drop),
            nn.Linear(128,64),     nn.ReLU(), nn.Dropout(p_drop),
            nn.Linear(64, n_classes))
    def forward(self, x): return self.net(x)

def pgd(model, x, y, eps=0.01, alpha=0.004, steps=7):
    model.eval(); x0 = x.detach(); x_adv = x0.clone().detach()
    for _ in range(steps):
        x_adv.requires_grad_(True)
        loss = F.cross_entropy(model(x_adv), y)
        grad = torch.autograd.grad(loss, x_adv, retain_graph=False, create_graph=False)[0]
        x_adv = x_adv.detach() + alpha * grad.sign()
        x_adv = torch.max(torch.min(x_adv, x0 + eps), x0 - eps)
    return x_adv.detach()

log('Attack helpers defined.')
log('MLP baseline defined.')
print("Cell 4 is completed")
```

```python
# ── Cell 5 — UVIF-IoT architecture: evidential, gated, Lagrangian ────────────
print("Cell 5: UVIF architecture entered")

# ── 5.1  Evidential MLP (§III-D) ─────────────────────────────────────────────
class EvidentialMLP(nn.Module):
    def __init__(self, in_dim, n_classes, p_drop=0.3):
        super().__init__()
        self.fc1  = nn.Linear(in_dim, 256)
        self.fc2  = nn.Linear(256, 128)
        self.fc3  = nn.Linear(128, 64)
        self.fc4  = nn.Linear(64, n_classes)
        self.drop = nn.Dropout(p=p_drop)
        self.n_cls = n_classes
    def forward(self, x):
        h = self.drop(F.relu(self.fc1(x)))
        h = self.drop(F.relu(self.fc2(h)))
        h = self.drop(F.relu(self.fc3(h)))
        return self.fc4(h)
    def evidential(self, x):
        z     = self.forward(x)
        alpha = torch.exp(z.clamp(-10, 10)) + 1.0
        S     = alpha.sum(dim=1, keepdim=True)
        p_hat = alpha / S
        psi_S = torch.digamma(S + 1)
        psi_a = torch.digamma(alpha + 1)
        u_epi = (p_hat * (psi_S - psi_a)).sum(dim=1)
        u_ale = -(p_hat * (p_hat + 1e-8).log()).sum(dim=1)
        return alpha, p_hat, u_epi, u_ale

def evidential_loss(alpha, y, n_classes, beta=0.10):
    S        = alpha.sum(dim=1, keepdim=True)
    p_hat    = alpha / S
    one_hot  = F.one_hot(y, num_classes=n_classes).float()
    brier    = ((one_hot - p_hat) ** 2 + p_hat * (1.0 - p_hat) / (S + 1.0)).sum(dim=1).mean()
    alpha_t  = one_hot + (1.0 - one_hot) * alpha
    S_tilde  = alpha_t.sum(dim=1)
    # FIX: .to(alpha.device) prevents CPU/GPU device mismatch that caused CUDA assert
    kl = (torch.lgamma(S_tilde)
          - torch.lgamma(torch.tensor(float(n_classes)).to(alpha.device))
          - torch.lgamma(alpha_t).sum(dim=1)
          + ((alpha_t - 1.0) * (torch.digamma(alpha_t)
             - torch.digamma(S_tilde.unsqueeze(1)))).sum(dim=1)).mean()
    return brier + beta * kl

# ── 5.2  Energy Gate (§III-F) — Concrete/Gumbel-sigmoid ─────────────────────
class EnergyGate(nn.Module):
    def __init__(self, in_dim, hidden=64):
        super().__init__()
        self.gate = nn.Sequential(nn.Linear(in_dim, hidden), nn.ReLU(), nn.Linear(hidden, in_dim))
    def forward(self, x, tau=1.0, hard=False):
        logits = self.gate(x)
        noise  = -torch.log(-torch.log(torch.rand_like(logits).clamp(1e-8, 1.0 - 1e-8)))
        soft   = torch.sigmoid((logits + noise) / tau)
        if hard:
            return (soft >= 0.5).float() - soft.detach() + soft   # straight-through
        return soft
    def energy_cost(self, mask, cost_vec):
        c = torch.tensor(cost_vec, dtype=torch.float32, device=mask.device)
        return (mask * c).sum(dim=1).mean()
    def inclusion_probs(self, x, tau=0.1, n_samples=5):
        masks = torch.stack([self.forward(x, tau=tau, hard=False) for _ in range(n_samples)])
        return masks.mean(dim=0)

# ── 5.3  Dual Optimizer (§III-B) ─────────────────────────────────────────────
class DualOptimizer:
    def __init__(self, cfg):
        self.eta    = cfg.get('eta_mu', 0.03)
        self.mu_max = cfg.get('mu_max', 2.0)
        self.eps_r  = cfg.get('epsilon_r', 0.20)
        self.eps_c  = cfg.get('epsilon_c', 0.15)
        self.eps_e  = cfg.get('epsilon_e', 0.35)
        self.warmup = cfg.get('mu_warmup_epochs', 8)
        self.mu     = [0.0, 0.0, 0.0]
    def step(self, R_dro, ece, energy, epoch):
        if epoch < self.warmup: return
        violations = [float(R_dro)-self.eps_r, float(ece)-self.eps_c, float(energy)-self.eps_e]
        for i, v in enumerate(violations):
            self.mu[i] = float(np.clip(self.mu[i] + self.eta * v, 0.0, self.mu_max))
    @property
    def mu1(self): return self.mu[0]
    @property
    def mu2(self): return self.mu[1]
    @property
    def mu3(self): return self.mu[2]
    def state_dict(self): return {'mu': list(self.mu)}
    def load_state_dict(self, d): self.mu = list(d.get('mu', [0.0, 0.0, 0.0]))

# ── 5.4  Brier calibration regulariser (§III-E) ──────────────────────────────
def brier_calibration_loss(probs, y, beta_cal=0.20):
    one_hot  = F.one_hot(y, num_classes=probs.shape[1]).float()
    brier    = ((probs - one_hot) ** 2).sum(dim=1).mean()
    # FIX: renormalise p_bar and correct KL direction: KL(p_bar || probs)
    p_bar    = probs.detach().mean(dim=0, keepdim=True).clamp(1e-8, 1.0)
    p_bar    = p_bar / p_bar.sum(dim=1, keepdim=True)
    probs_c  = probs.clamp(1e-8, 1.0)
    kl_sharp = F.kl_div(probs_c.log(), p_bar.expand_as(probs_c), reduction='batchmean')
    return brier + beta_cal * kl_sharp

# ── 5.5  Mahalanobis-FGSM placeholder (§III-C) ───────────────────────────────
# Full covariance-based version requires init_mahalanobis(); this is a stable
# signed-gradient fallback that never conflicts with the Cell 4 definition.
def mahalanobis_fgsm(model, x, y, rho=0.01, ds_key=None):
    x_adv = x.detach().clone().requires_grad_(True)
    loss  = F.cross_entropy(model(x_adv), y)
    grad  = torch.autograd.grad(loss, x_adv, retain_graph=False, create_graph=False)[0]
    return (x_adv + rho * grad.sign()).detach()

def init_mahalanobis(X, ds_key='default'):
    return None   # placeholder; extend with real Sigma_inv for final run

# ── 5.6  UVIF Loss (§III-B/C/E) ──────────────────────────────────────────────
class UVIFLoss(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.beta_evid = cfg.get('beta_evid', 0.10)
        self.beta_cal  = cfg.get('beta_cal',  0.20)
        self.rho       = cfg.get('rho_wass',  0.01)

    def _batch_ece(self, probs, y):
        # FIX: MSE proxy — avoids F.binary_cross_entropy CUDA assert on NaN inputs
        conf = probs.max(dim=1).values.clamp(0.0, 1.0)
        corr = (probs.argmax(dim=1) == y).float()
        return ((conf - corr) ** 2).mean()

    def forward(self, evid_mlp, gate, x, y, tau, cost_vec, dual_opt=None,
                class_weights=None, ds_key=None, epoch=0, warmup=8):
        x, y = x.to(DEVICE), y.to(DEVICE)
        cw   = class_weights.to(DEVICE) if class_weights is not None else None

        # ── Warmup: train evidential backbone only, no gating or DRO ─────────
        if epoch < warmup:
            alpha, p_hat, _, _ = evid_mlp.evidential(x)
            loss = evidential_loss(alpha, y, evid_mlp.n_cls, beta=self.beta_evid)
            # FIX: no redundant F.cross_entropy on top of evidential_loss
            return loss, {'evid': float(loss.item()), 'dro': 0.0, 'energy': 0.0,
                          'cal': 0.0, 'ece_proxy': 0.0, 'mu1': 0.0, 'mu2': 0.0, 'mu3': 0.0}

        # ── Post-warmup: full Lagrangian ──────────────────────────────────────
        mask   = gate(x, tau=tau, hard=False)
        x_gate = x * mask
        alpha, p_hat, _, _ = evid_mlp.evidential(x_gate)
        l_evid = evidential_loss(alpha, y, evid_mlp.n_cls, beta=self.beta_evid)
        # FIX: stable per-sample class weighting (no redundant cross-entropy)
        if cw is not None:
            l_evid = l_evid * cw[y].mean()

        x_adv           = mahalanobis_fgsm(evid_mlp, x_gate.detach(), y, rho=self.rho, ds_key=ds_key)
        alpha_adv, _, _, _ = evid_mlp.evidential(x_adv)
        l_dro    = evidential_loss(alpha_adv, y, evid_mlp.n_cls, beta=self.beta_evid)
        pi       = gate.inclusion_probs(x, tau=tau, n_samples=3)
        l_energy = gate.energy_cost(pi, cost_vec)
        l_cal    = brier_calibration_loss(p_hat, y, beta_cal=self.beta_cal)
        with torch.no_grad():
            ece_proxy = float(self._batch_ece(p_hat.detach(), y).item())

        if dual_opt is not None:
            mu1, mu2, mu3 = dual_opt.mu1, dual_opt.mu2, dual_opt.mu3
            total = (l_evid
                     + mu1 * (l_dro    - dual_opt.eps_r)
                     + mu2 * (l_cal    - dual_opt.eps_c)
                     + mu3 * (l_energy - dual_opt.eps_e))
        else:
            mu1 = mu2 = mu3 = 0.0
            total = (l_evid
                     + CFG.get('lambda_dro',0.05) * l_dro
                     + CFG.get('lambda_cal',0.03) * l_cal
                     + CFG.get('lambda_e',  0.01) * l_energy)

        return total, {'evid': float(l_evid.item()), 'dro': float(l_dro.item()),
                       'energy': float(l_energy.item()), 'cal': float(l_cal.item()),
                       'ece_proxy': ece_proxy, 'mu1': float(mu1), 'mu2': float(mu2), 'mu3': float(mu3)}

# ── 5.7  UVIF-IoT model (§III-G) ─────────────────────────────────────────────
class UVIFIoT(nn.Module):
    def __init__(self, in_dim, n_classes, cfg):
        super().__init__()
        self.evid  = EvidentialMLP(in_dim, n_classes, p_drop=cfg.get('p_drop', 0.3))
        self.gate  = EnergyGate(in_dim)
        self.n_cls = n_classes
    def forward(self, x, tau=0.1, hard=False):
        mask = self.gate(x, tau=tau, hard=hard)
        alpha, p_hat, u_epi, u_ale = self.evid.evidential(x * mask)
        return alpha, p_hat, u_epi, u_ale, mask
    def predict(self, x, tau=0.1):
        x    = x.to(DEVICE)
        mask = self.gate(x, tau=tau, hard=True)
        alpha, p_hat, u_epi, u_ale = self.evid.evidential(x * mask)
        return p_hat, u_epi, u_ale, mask
    @property
    def bnn(self): return self.evid   # backward-compat shim

log('UVIF-IoT architecture defined: Evidential + Energy Gate + Dual Lagrangian.')
print("Cell 5 is completed")
```

```python
# ── Cell 6 — Training and checkpointing: robust, global-state safe ───────────
print("Cell 6: training entered")

from sklearn.utils.class_weight import compute_class_weight
if torch.cuda.is_available(): torch.cuda.empty_cache()
gc.collect()

def eval_f1(model, loader, S=None):
    model.train() if S else model.eval()
    preds, targets = [], []
    with torch.no_grad():
        for x, y in loader:
            x = x.to(DEVICE)
            p = (torch.stack([F.softmax(model(x),-1) for _ in range(S)]).mean(0)
                 if S else F.softmax(model(x),-1))
            preds.extend(p.argmax(1).cpu().numpy()); targets.extend(y.numpy())
    return f1_score(targets, preds, average='macro', zero_division=0)

def eval_uvif_f1(uvif, loader, S=5, tau=0.1):
    uvif.eval(); preds, targets = [], []
    with torch.no_grad():
        for x, y in loader:
            p_hat, _, _, _ = uvif.predict(x.to(DEVICE), tau=tau)
            preds.extend(p_hat.argmax(1).cpu().numpy()); targets.extend(y.numpy())
    return f1_score(targets, preds, average='macro', zero_division=0)

def compute_torch_class_weights(y):
    if not CFG.get('use_class_weights', True): return None
    classes = np.unique(y)
    weights = compute_class_weight(class_weight='balanced', classes=classes, y=y)
    out = np.ones(int(classes.max())+1, dtype=np.float32)
    for c, w in zip(classes, weights): out[int(c)] = float(w)
    return torch.tensor(out, dtype=torch.float32, device=DEVICE)

def train_erm(model, tr_ldr, val_ldr, cfg, adversarial=False, tag='', class_weights=None):
    model.to(DEVICE)
    opt = torch.optim.Adam(model.parameters(), lr=cfg['lr'], weight_decay=cfg.get('weight_decay',0.0))
    hist = []; cw = class_weights.to(DEVICE) if class_weights is not None else None
    for ep in range(cfg['epochs']):
        model.train()
        for x, y in tr_ldr:
            x, y = x.to(DEVICE), y.to(DEVICE)
            if adversarial: x = pgd(model, x, y, cfg.get('rho_wass',0.01), cfg.get('pgd_alpha',0.004), 5)
            opt.zero_grad(set_to_none=True)
            F.cross_entropy(model(x), y, weight=cw).backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=5.0)
            opt.step()
        vf1 = eval_f1(model, val_ldr); hist.append(vf1)
        if (ep+1) % 10 == 0 or ep == 0:
            log(f'[{tag}] ep {ep+1}/{cfg["epochs"]}  val_f1={vf1:.4f}')
    return hist

def train_uvif(uvif, tr_ldr, val_ldr, cfg, cost_vec, tag='UVIF-IoT', class_weights=None, ds_key=None):
    uvif.to(DEVICE)
    crit     = UVIFLoss(cfg)
    dual_opt = DualOptimizer(cfg)
    opt      = torch.optim.Adam(uvif.parameters(), lr=cfg['lr'], weight_decay=cfg.get('weight_decay',0.0))
    tau_sch  = np.linspace(cfg.get('tau_start',1.0), cfg.get('tau_end',0.05), cfg['epochs'])
    warmup   = cfg.get('uvif_warmup_epochs', 8)
    patience = cfg.get('early_stop_patience', cfg['epochs'])
    hist, ch = [], []; best_f1, best_state, wait = -1.0, None, 0
    for ep in range(cfg['epochs']):
        uvif.train(); tau = float(tau_sch[ep]); tot = 0.0
        comp = {k:0.0 for k in ['evid','dro','energy','cal','ece_proxy','mu1','mu2','mu3']}
        for x, y in tr_ldr:
            opt.zero_grad(set_to_none=True)
            loss, c = crit(uvif.evid, uvif.gate, x, y, tau, cost_vec,
                           dual_opt=dual_opt, class_weights=class_weights,
                           ds_key=ds_key, epoch=ep, warmup=warmup)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(uvif.parameters(), max_norm=5.0)
            opt.step(); tot += float(loss.item())
            for k in comp: comp[k] += float(c.get(k, 0.0))
        n = max(1, len(tr_ldr))
        dual_opt.step(comp['dro']/n, comp['ece_proxy']/n, comp['energy']/n, ep)
        vf1 = eval_uvif_f1(uvif, val_ldr, S=cfg.get('mc_samples',5), tau=tau)
        hist.append(vf1)
        ch.append({'epoch':ep+1,'tau':tau,'loss':tot/n, **{k:v/n for k,v in comp.items()},
                   'mu1_final':dual_opt.mu1,'mu2_final':dual_opt.mu2,'mu3_final':dual_opt.mu3})
        if (ep+1) % 5 == 0 or ep == 0:
            log(f'[{tag}] ep {ep+1}/{cfg["epochs"]}  tau={tau:.3f}  loss={tot/n:.4f}  val_f1={vf1:.4f}  '
                f'mu=({dual_opt.mu1:.3f},{dual_opt.mu2:.3f},{dual_opt.mu3:.3f})')
        if vf1 > best_f1 + 1e-4:
            best_f1 = vf1; best_state = {k:v.detach().cpu().clone() for k,v in uvif.state_dict().items()}; wait = 0
        else:
            wait += 1
            if wait >= patience and ep+1 > warmup:
                log(f'[{tag}] Early stopping at epoch {ep+1} (best val_f1={best_f1:.4f})'); break
    if best_state is not None: uvif.load_state_dict(best_state)
    return hist, ch, dual_opt

MODELS = {}; HISTORIES = {}
DATASET_REGISTRY = [
    ('nbaiot', nb_loaders, X_nb_train, y_nb_train, X_nb_test, y_nb_test),
    ('unsw',   unsw_loaders, X_unsw_train, y_unsw_train, X_unsw_test, y_unsw_test),
]

for ds, loaders, X_tr, y_tr, X_te, y_te in DATASET_REGISTRY:
    tr_ldr, val_ldr, te_ldr = loaders
    in_d, n_cls = X_tr.shape[1], len(np.unique(y_tr))
    MODELS[ds], HISTORIES[ds] = {}, {}
    cost_vec     = np.ones(in_d, dtype=np.float32) / in_d
    class_weights = compute_torch_class_weights(y_tr)
    if class_weights is not None:
        log(f'[{ds}] class weights: {class_weights.detach().cpu().numpy().round(3).tolist()}')
    log(f'\n=== {ds.upper()} | in_dim={in_d} | n_classes={n_cls} ===')
    for tag, p_drop, adv in [('ERM-MLP',0.0,False),('AT-ERM',0.0,True),('MC-Dropout',CFG.get('p_drop',0.3),False)]:
        model = MLP(in_d, n_cls, p_drop=p_drop)
        hist  = train_erm(model, tr_ldr, val_ldr, CFG, adversarial=adv, tag=tag, class_weights=class_weights)
        torch.save(model.state_dict(), CKPT_DIR / f'{tag.lower().replace("-","_")}_{ds}.pt')
        MODELS[ds][tag], HISTORIES[ds][tag] = model, hist
        gc.collect();
        if torch.cuda.is_available(): torch.cuda.empty_cache()
    if not CFG.get('skip_rf_for_speed', False):
        try:
            rf = RandomForestClassifier(n_estimators=CFG.get('rf_n_estimators',80), n_jobs=-1,
                                        random_state=SEED, class_weight='balanced', max_samples=0.7, bootstrap=True)
            rf.fit(X_tr, y_tr); MODELS[ds]['Random Forest'] = rf; log(f'[{ds}] Random Forest trained.')
        except Exception as e: log(f'[{ds}] Random Forest skipped: {e}')
    try:
        iso = IsolationForest(n_estimators=80, contamination='auto', random_state=SEED, max_samples=min(30000,len(X_tr)))
        iso.fit(X_tr); MODELS[ds]['Isolation Forest'] = iso; log(f'[{ds}] Isolation Forest trained.')
    except Exception as e: log(f'[{ds}] Isolation Forest skipped: {e}')
    uvif = UVIFIoT(in_d, n_cls, CFG)
    try:
        benign_mask = (y_tr == 0)
        init_mahalanobis(X_tr[benign_mask] if benign_mask.sum() > 100 else X_tr, ds_key=ds)
    except Exception: pass
    h_u, h_c, dual_saved = train_uvif(uvif, tr_ldr, val_ldr, CFG, cost_vec, class_weights=class_weights, ds_key=ds)
    torch.save(uvif.state_dict(), CKPT_DIR / f'uvif_{ds}.pt')
    MODELS[ds]['UVIF-IoT'] = uvif
    HISTORIES[ds]['UVIF-IoT'] = (h_u, h_c)
    HISTORIES[ds]['UVIF-IoT-dual'] = dual_saved
    log(f'[{ds}] models stored: {list(MODELS[ds].keys())}')
    gc.collect();
    if torch.cuda.is_available(): torch.cuda.empty_cache()

log('All models trained and checkpointed.')
for ds in ['nbaiot','unsw']:
    rows = {}
    for nm in ['ERM-MLP','AT-ERM','MC-Dropout']:
        if nm in HISTORIES.get(ds,{}): rows[nm] = list(HISTORIES[ds][nm])
    if 'UVIF-IoT' in HISTORIES.get(ds,{}):
        h_u, _ = HISTORIES[ds]['UVIF-IoT']; rows['UVIF-IoT'] = list(h_u)
    if rows:
        max_len = max(len(v) for v in rows.values())
        hist_df = pd.DataFrame({k: v+[np.nan]*(max_len-len(v)) for k,v in rows.items()})
        hist_df.index = np.arange(1, max_len+1); hist_df.index.name = 'epoch'
        hist_df.to_csv(TABLE_DIR / f'val_f1_history_{ds}.csv')
        log(f'[{ds}] Validation F1 history saved.')

if len(MODELS) == 0: raise RuntimeError('MODELS is empty after training.')
print('MODELS keys:', MODELS.keys())
for ds in MODELS: print(ds, 'models:', MODELS[ds].keys())
print('Cell 6 is completed')
```

```python
# ── Cell 7 — Clean evaluation and calibration metrics ───────────────────────
print("Cell 7: clean evaluation entered")

if 'MODELS' not in globals() or len(MODELS) == 0:
    raise RuntimeError('MODELS missing. Run Cell 6 first.')

CLEAN = {}; PROBS = {}; TARGETS = {}; PREDS = {}

def collect_probs(model, loader=None, S=None, is_rf=False, is_iso=False,
                  is_uvif=False, X_test=None, y_test=None):
    if is_rf:
        p = model.predict_proba(X_test)
        return p, p.argmax(axis=1), y_test
    if is_iso:
        raw  = model.predict(X_test)
        sc   = -model.score_samples(X_test)
        asc  = (sc - sc.min()) / (sc.max() - sc.min() + 1e-8)
        if len(np.unique(y_test)) > 2:
            p = np.zeros((len(y_test), len(np.unique(y_test))), dtype=float)
            p[:,0] = 1.0-asc
            if p.shape[1] > 1: p[:,1] = asc
            return p, (raw==-1).astype(int), y_test
        p = np.vstack([1.0-asc, asc]).T
        return p, (raw==-1).astype(int), y_test
    if is_uvif:
        model.eval(); all_p=[]; all_pred=[]; all_tgt=[]
        with torch.no_grad():
            for x, y in loader:
                p_hat,_,_,_ = model.predict(x, tau=CFG.get('tau_end',0.05))
                all_p.append(p_hat.cpu())
                all_pred.extend(p_hat.argmax(1).cpu().numpy()); all_tgt.extend(y.numpy())
        probs = torch.cat(all_p).numpy()
        return probs, np.array(all_pred), np.array(all_tgt)
    model.train() if S else model.eval()
    all_p=[]; all_pred=[]; all_tgt=[]
    with torch.no_grad():
        for x, y in loader:
            x = x.to(DEVICE)
            p = (torch.stack([F.softmax(model(x),-1) for _ in range(S)]).mean(0) if S
                 else F.softmax(model(x),-1))
            all_p.append(p.cpu()); all_pred.extend(p.argmax(1).cpu().numpy()); all_tgt.extend(y.numpy())
    return torch.cat(all_p).numpy(), np.array(all_pred), np.array(all_tgt)

def expected_calibration_error(confs, correct, n_bins=15):
    bins = np.linspace(0,1,n_bins+1); ece=0.0; mce=0.0
    for lo, hi in zip(bins[:-1], bins[1:]):
        mask = (confs>lo)&(confs<=hi)
        if not np.any(mask): continue
        gap = abs(correct[mask].mean() - confs[mask].mean())
        ece += mask.mean()*gap; mce = max(mce, gap)
    return float(ece), float(mce)

def safe_auc_metrics(targets, probs):
    out = {'ROC_AUC': np.nan, 'PR_AUC': np.nan}
    try:
        if probs.shape[1] == 2:
            out['ROC_AUC'] = roc_auc_score(targets, probs[:,1])
            out['PR_AUC']  = average_precision_score(targets, probs[:,1])
        else:
            out['ROC_AUC'] = roc_auc_score(targets, probs, multi_class='ovr', average='macro')
            out['PR_AUC']  = average_precision_score(np.eye(probs.shape[1])[targets], probs, average='macro')
    except Exception: pass
    return out

for ds, loaders, X_te, y_te in [
    ('nbaiot', nb_loaders, X_nb_test, y_nb_test),
    ('unsw',   unsw_loaders, X_unsw_test, y_unsw_test),
]:
    _,_,te_ldr = loaders
    CLEAN[ds]={}; PROBS[ds]={}; TARGETS[ds]={}; PREDS[ds]={}
    M=MODELS[ds]; S=CFG.get('mc_samples',5)
    specs = []
    if 'ERM-MLP'         in M: specs.append(('ERM-MLP',         M['ERM-MLP'],         te_ldr, None, False, False, False))
    if 'AT-ERM'          in M: specs.append(('AT-ERM',          M['AT-ERM'],          te_ldr, None, False, False, False))
    if 'MC-Dropout'      in M: specs.append(('MC-Dropout',      M['MC-Dropout'],      te_ldr, S,    False, False, False))
    if 'Random Forest'   in M: specs.append(('Random Forest',   M['Random Forest'],   None,   None, True,  False, False))
    if 'Isolation Forest'in M: specs.append(('Isolation Forest',M['Isolation Forest'],None,   None, False, True,  False))
    if 'UVIF-IoT'        in M: specs.append(('UVIF-IoT',        M['UVIF-IoT'],        te_ldr, S,    False, False, True))
    for name,model,ldr,mc,is_rf,is_iso,is_uvif in specs:
        probs,preds,targets = collect_probs(model,loader=ldr,S=mc,is_rf=is_rf,is_iso=is_iso,
                                            is_uvif=is_uvif,X_test=X_te,y_test=y_te)
        confs  = probs.max(axis=1); correct = (targets==preds).astype(int)
        ece,mce= expected_calibration_error(confs, correct, n_bins=CFG.get('ece_bins',15))
        row    = dict(F1=f1_score(targets,preds,average='macro',zero_division=0),
                      Acc=accuracy_score(targets,preds), BalAcc=balanced_accuracy_score(targets,preds),
                      Precision=precision_score(targets,preds,average='macro',zero_division=0),
                      Recall=recall_score(targets,preds,average='macro',zero_division=0),
                      ECE=ece, MCE=mce, **safe_auc_metrics(targets,probs))
        CLEAN[ds][name]=row; PROBS[ds][name]=probs; TARGETS[ds][name]=targets; PREDS[ds][name]=preds
        log(f'[{ds}] {name:22s}  F1={row["F1"]:.4f}  Acc={row["Acc"]:.4f}  BalAcc={row["BalAcc"]:.4f}  ECE={row["ECE"]:.4f}')
    pd.DataFrame(CLEAN[ds]).T.to_csv(TABLE_DIR / f'clean_metrics_{ds}.csv')
    log(f'[{ds}] Clean metrics saved.')
print('Cell 7 is completed')
```

```python
# ── Cell 8 — Figures: reliability, loss components, convergence ──────────────
print("Cell 8: figures entered")
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
for ax, ds in zip(axes, ['nbaiot','unsw']):
    if ds not in PROBS:
        ax.text(0.5,0.5,'Not available',ha='center',va='center',transform=ax.transAxes); continue
    for name in ['ERM-MLP','MC-Dropout','UVIF-IoT']:
        if name not in PROBS[ds]: continue
        probs=PROBS[ds][name]; preds=PREDS[ds][name]; targets=TARGETS[ds][name]
        confs=probs.max(axis=1); correct=(targets==preds).astype(int)
        try:
            pt, pp = calibration_curve(correct, confs, n_bins=CFG.get('ece_bins',15), strategy='uniform')
            ax.plot(pp, pt, marker='o', label=name)
        except Exception: pass
    ax.plot([0,1],[0,1],'k--',label='Perfect calibration')
    ax.set_xlabel('Mean confidence'); ax.set_ylabel('Observed accuracy')
    ax.set_title(f'Reliability — {ds.upper()}'); ax.legend(fontsize=8)
plt.tight_layout()
plt.savefig(FIG_DIR/'reliability_diagrams.pdf', dpi=300, bbox_inches='tight'); plt.show()
log('Reliability diagrams saved.')

component_candidates = {'Evidence / ELBO':['elbo','evid','loss'],'DRO':['dro'],'Energy':['energy'],'Calibration':['cal','ece_proxy']}
fig, axes = plt.subplots(2,2,figsize=(12,8))
for ax,(label,keys) in zip(axes.flat, component_candidates.items()):
    plotted = False
    for ds in ['nbaiot','unsw']:
        if ds not in HISTORIES or 'UVIF-IoT' not in HISTORIES[ds]: continue
        _, ch = HISTORIES[ds]['UVIF-IoT']
        if not ch: continue
        selected = next((k for k in keys if isinstance(ch[0],dict) and k in ch[0]), None)
        if selected is None: continue
        ax.plot([c.get(selected,np.nan) for c in ch], label=ds.upper()); plotted=True
    ax.set_title(label); ax.set_xlabel('Epoch'); ax.set_ylabel('Value')
    if plotted: ax.legend(fontsize=8)
    else: ax.text(0.5,0.5,'Not available',ha='center',va='center',transform=ax.transAxes)
plt.suptitle('UVIF-IoT Loss Component Trajectories')
plt.tight_layout(); plt.savefig(FIG_DIR/'loss_components.pdf', dpi=300, bbox_inches='tight'); plt.show()

fig, axes = plt.subplots(1,2,figsize=(12,4))
for ax, ds in zip(axes, ['nbaiot','unsw']):
    if ds not in HISTORIES: continue
    for nm in ['ERM-MLP','AT-ERM','MC-Dropout']:
        if nm in HISTORIES[ds]: ax.plot(HISTORIES[ds][nm], label=nm)
    if 'UVIF-IoT' in HISTORIES[ds]:
        h, _ = HISTORIES[ds]['UVIF-IoT']; ax.plot(h,label='UVIF-IoT',linewidth=2,linestyle='--')
    ax.set_xlabel('Epoch'); ax.set_ylabel('Validation macro-F1')
    ax.set_title(f'Convergence — {ds.upper()}'); ax.legend(fontsize=8)
plt.tight_layout(); plt.savefig(FIG_DIR/'training_convergence.pdf', dpi=300, bbox_inches='tight'); plt.show()
log('Training figures saved.')
print("Cell 8 is completed")
```

```python
# ── Cell 9 — Adversarial robustness evaluation ───────────────────────────────
print("Cell 9: adversarial evaluation entered")

NN_NAMES = ['ERM-MLP','AT-ERM','MC-Dropout','UVIF-IoT']
ADV = {}

# FIX: composed gate+evid forward — no gate bypass for UVIF-IoT
def _get_eval_model(ds, nm):
    if nm == 'UVIF-IoT':
        uvif_inst = MODELS[ds][nm]
        class _GatedForward(nn.Module):
            def forward(self_, x):
                x    = x.to(DEVICE)
                mask = uvif_inst.gate(x, tau=CFG.get('tau_end',0.05), hard=True)
                return uvif_inst.evid(x * mask)
        return _GatedForward().to(DEVICE)
    return MODELS[ds][nm]

def f1_under_attack(model, loader, eps=0.01, pgd_steps=1, batch_cap=50):
    preds, targets = [], []; model.eval()
    for bi, (x, y) in enumerate(loader):
        if batch_cap is not None and bi >= batch_cap: break
        x, y = x.to(DEVICE), y.to(DEVICE)
        x_adv = pgd(model, x, y, eps=eps, alpha=CFG.get('pgd_alpha',0.004), steps=pgd_steps)
        with torch.no_grad():
            p = F.softmax(model(x_adv), dim=-1)
        preds.extend(p.argmax(1).cpu().numpy()); targets.extend(y.cpu().numpy())
    return f1_score(targets, preds, average='macro', zero_division=0)

cap = CFG.get('adv_batch_cap', 50)
for ds, loaders in [('nbaiot', nb_loaders), ('unsw', unsw_loaders)]:
    _, _, te_ldr = loaders; ADV[ds] = {}
    log(f'[{ds}] Starting adversarial evaluation. Batch cap: {cap}')
    for nm in NN_NAMES:
        if nm not in MODELS.get(ds, {}): continue
        m = _get_eval_model(ds, nm); ADV[ds][nm] = {}
        for eps, label in [(0.01,'e01'),(0.02,'e02')]:
            ADV[ds][nm][f'FGSM_{label}']               = f1_under_attack(m, te_ldr, eps=eps, pgd_steps=1,                           batch_cap=cap)
            ADV[ds][nm][f'PGD{CFG.get("pgd_steps",10)}_{label}'] = f1_under_attack(m, te_ldr, eps=eps, pgd_steps=CFG.get('pgd_steps',10), batch_cap=cap)
        log(f'[{ds}] {nm:12s}  FGSM_e01={ADV[ds][nm]["FGSM_e01"]:.4f}  '
            f'PGD_e02={ADV[ds][nm][f"PGD{CFG.get(chr(39)+ "pgd_steps"+ chr(39),10)}_e02"]:.4f}')
    pd.DataFrame(ADV[ds]).T.to_csv(TABLE_DIR / f'adversarial_metrics_{ds}.csv')
    log(f'[{ds}] Adversarial metrics saved.')
log('Adversarial evaluation complete.')
print('Cell 9 is completed')
```

```python
# ── Cell 10 — Energy-gated inference and saliency fidelity ───────────────────
print("Cell 10: energy/saliency entered")

GATE_RESULTS = {}; fid_nb, fid_unsw = [], []

def eval_uvif_with_budget(uvif, loader, budget_frac=0.25):
    uvif.eval(); preds=[]; targets=[]; feat_counts=[]
    with torch.no_grad():
        for x, y in loader:
            x = x.to(DEVICE)
            mask = uvif.gate(x, tau=CFG.get('tau_end',0.05), hard=False)
            k    = max(1, int(mask.shape[1]*budget_frac))
            idx  = torch.topk(mask, k=k, dim=1).indices
            hard = torch.zeros_like(mask).scatter_(1, idx, 1.0)
            p    = uvif.evid.evidential(x*hard)[1]
            preds.extend(p.argmax(1).cpu().numpy()); targets.extend(y.numpy())
            feat_counts.extend(hard.sum(1).cpu().numpy())
    f1          = f1_score(targets, preds, average='macro', zero_division=0)
    avg_features= float(np.mean(feat_counts))
    saving      = (1 - avg_features/mask.shape[1]) * 100
    return f1, avg_features, saving

profiles = {'Cortex-M4 @ 168 MHz':0.22, 'Cortex-M7 @ 216 MHz':0.30, 'ESP32 @ 240 MHz':0.26}
for ds, loaders in [('nbaiot',nb_loaders),('unsw',unsw_loaders)]:
    GATE_RESULTS[ds] = {}
    if 'UVIF-IoT' not in MODELS.get(ds,{}): continue
    _,_,te_ldr = loaders; uvif = MODELS[ds]['UVIF-IoT']
    for prof, budget in profiles.items():
        f1,feat,sav = eval_uvif_with_budget(uvif, te_ldr, budget_frac=budget)
        GATE_RESULTS[ds][prof] = {'F1':f1,'avg_features':feat,'saving_pct':sav}
        log(f'[{ds}] {prof} budget={budget:.2f} F1={f1:.4f} features={feat:.1f} saving={sav:.1f}%')

def saliency_fidelity(ds, X_test, y_test, topks=[5,10,20,30,50], max_samples=5000):
    if 'UVIF-IoT' not in MODELS.get(ds,{}): return []
    uvif = MODELS[ds]['UVIF-IoT']; uvif.eval()
    X    = torch.tensor(X_test[:max_samples], dtype=torch.float32).to(DEVICE)
    y    = y_test[:max_samples]
    with torch.no_grad():
        mask = uvif.gate.inclusion_probs(X, tau=CFG.get('tau_end',0.05), n_samples=3)
        sal  = mask.mean(0).cpu().numpy()
    order = np.argsort(-sal); out = []
    for k in topks:
        X_ab = X_test[:max_samples].copy(); X_ab[:, order[:min(k,X_ab.shape[1])]] = 0.0
        ldr  = make_loader(X_ab, y, shuffle=False)
        f1   = eval_uvif_f1(uvif, ldr, S=CFG.get('mc_samples',5), tau=CFG.get('tau_end',0.05))
        out.append((k,f1)); log(f'[{ds}] Top-{k:3d} ablated: F1={f1:.4f}')
    return out

fid_nb   = saliency_fidelity('nbaiot', X_nb_test,   y_nb_test)
fid_unsw = saliency_fidelity('unsw',   X_unsw_test, y_unsw_test)
pd.DataFrame([(k,v) for k,v in fid_nb],   columns=['top_k','f1']).to_csv(TABLE_DIR/'saliency_fidelity_nbaiot.csv', index=False)
pd.DataFrame([(k,v) for k,v in fid_unsw], columns=['top_k','f1']).to_csv(TABLE_DIR/'saliency_fidelity_unsw.csv',  index=False)
log('Energy-gated inference and saliency fidelity complete.')
print('Cell 10 is completed')
```

```python
# ── Cell 11 — Optional light Pareto analysis ─────────────────────────────────
print("Cell 11: Pareto analysis entered")
import matplotlib.pyplot as plt

PARETO = pd.DataFrame()
if CFG.get('run_pareto',True) and 'unsw' in MODELS and 'UVIF-IoT' in MODELS['unsw']:
    rows = []; _,_,te_ldr = unsw_loaders
    for budget in [0.15,0.22,0.30,0.45,0.60,1.00]:
        f1,feat,sav = eval_uvif_with_budget(MODELS['unsw']['UVIF-IoT'], te_ldr, budget_frac=budget)
        rows.append({'dataset':'unsw','budget_frac':budget,'F1':f1,'active_features':feat,'saving_pct':sav})
    PARETO = pd.DataFrame(rows)
    PARETO.to_csv(TABLE_DIR/'pareto_uvif_budget_unsw.csv', index=False)
    display(PARETO)
    plt.figure(figsize=(6,4))
    plt.plot(PARETO['active_features'], PARETO['F1'], marker='o')
    plt.xlabel('Active features'); plt.ylabel('Macro-F1')
    plt.title('UVIF-IoT Pareto Trade-off: Accuracy vs Feature Budget')
    plt.tight_layout(); plt.savefig(FIG_DIR/'pareto_tradeoff_unsw.pdf', dpi=300, bbox_inches='tight'); plt.show()
    log('Pareto analysis completed.')
else:
    log('Pareto analysis skipped.')
print('Cell 11 is completed')
```

```python
# ── Cell 12 — Tables, Q1 diagnosis, and final summary ────────────────────────
print("Cell 12: final outputs entered")

Q1_DIAGNOSIS = []
def add_diag(msg): Q1_DIAGNOSIS.append(msg); log('[Q1-DIAGNOSIS] '+msg)

if 'unsw' in CLEAN and 'UVIF-IoT' in CLEAN['unsw'] and 'MC-Dropout' in CLEAN['unsw']:
    diff = CLEAN['unsw']['UVIF-IoT']['F1'] - CLEAN['unsw']['MC-Dropout']['F1']
    if diff >= 0:
        add_diag(f'UNSW: UVIF-IoT improves clean F1 over MC-Dropout by {diff:.4f}.')
    else:
        add_diag(f'UNSW: MC-Dropout exceeds UVIF-IoT on clean F1 by {-diff:.4f}; frame UVIF-IoT as robustness/efficiency-oriented unless tuning closes the gap.')
for ds in ['nbaiot','unsw']:
    uv = CLEAN.get(ds,{}).get('UVIF-IoT',{})
    if uv.get('MCE',0) > 0.25:
        add_diag(f'{ds.upper()}: UVIF-IoT has high MCE={uv.get("MCE",np.nan):.4f}; include high-confidence calibration close-up and temperature scaling.')
    if ds in GATE_RESULTS:
        add_diag(f'{ds.upper()}: device-aware costs produce differentiated active-feature behavior across profiles.')

Q1_PATH.write_text('\n'.join(Q1_DIAGNOSIS), encoding='utf-8')

for ds in ['nbaiot','unsw']:
    if ds in CLEAN:
        pd.DataFrame(CLEAN[ds]).T.to_csv(TABLE_DIR / f'clean_metrics_{ds}.csv')
        pd.DataFrame(CLEAN[ds]).T.to_latex(TABLE_DIR / f'table_clean_{ds}.tex', float_format='%.4f')
    if ds in ADV:
        pd.DataFrame(ADV[ds]).T.to_csv(TABLE_DIR / f'adversarial_metrics_{ds}.csv')
        pd.DataFrame(ADV[ds]).T.to_latex(TABLE_DIR / f'table_adv_{ds}.tex', float_format='%.4f')
    if ds in GATE_RESULTS:
        pd.DataFrame(GATE_RESULTS[ds]).T.to_csv(TABLE_DIR / f'energy_gated_{ds}.csv')
        pd.DataFrame(GATE_RESULTS[ds]).T.to_latex(TABLE_DIR / f'table_energy_{ds}.tex', float_format='%.4f')

SUMMARY_PATH.write_text('', encoding='utf-8')
def s(msg=''): summarise(msg)
s('UVIF-IoT — Experimental Results Summary')
s('='*60); s(); s(f'[ Reportability ] REPORTABLE_RESULTS={REPORTABLE_RESULTS}')
if not REPORTABLE_RESULTS: s('  Synthetic smoke-test data detected. Do not use as manuscript evidence.')
s(); s('[ Clean-Traffic Detection ]'); s()
for ds in ['nbaiot','unsw']:
    s(f'  {ds.upper()}')
    for nm,r in CLEAN.get(ds,{}).items():
        s(f'    {nm:22s}  F1={r.get("F1",np.nan):.4f}  Acc={r.get("Acc",np.nan):.4f}  BalAcc={r.get("BalAcc",np.nan):.4f}  ECE={r.get("ECE",np.nan):.4f}  MCE={r.get("MCE",np.nan):.4f}')
s(); s('[ Adversarial Robustness ]'); s()
for ds in ['nbaiot','unsw']:
    s(f'  {ds.upper()}')
    for nm,row in ADV.get(ds,{}).items():
        s(f'    {nm:22s}  ' + '  '.join(f'{k}={row.get(k,np.nan):.4f}' for k in row))
s(); s('[ Energy-Gated Inference ]'); s()
for ds in ['nbaiot','unsw']:
    s(f'  {ds.upper()}')
    for prof,r in GATE_RESULTS.get(ds,{}).items():
        s(f'    {prof:25s}  F1={r.get("F1",np.nan):.4f}  features={r.get("avg_features",np.nan):.1f}  saving={r.get("saving_pct",np.nan):.1f}%')
s(); s('[ SALIENCY-IoT Fidelity Ablation ]'); s()
for ds, fid in [('nbaiot',fid_nb),('unsw',fid_unsw)]:
    s(f'  {ds.upper()}')
    for k, f1 in fid: s(f'    Top-{k:3d} ablated: F1={f1:.4f}')
s(); s('[ Q1-Readiness Diagnosis ]')
for msg in Q1_DIAGNOSIS: s('  - '+msg)
s(); s('[ Output Files ]'); s()
rows = []
for folder in [FIG_DIR, TABLE_DIR, OUTPUT_DIR, CKPT_DIR]:
    for fp in sorted(folder.glob('*')):
        if fp.is_file():
            rows.append({'category':folder.name,'filename':fp.name,
                         'size_kb':round(fp.stat().st_size/1024,1),'path':str(fp)})
            s(f'    [{folder.name}] {fp.name}  ({round(fp.stat().st_size/1024,1)} KB)')
inv = pd.DataFrame(rows); inv.to_csv(OUTPUT_DIR/'artifact_inventory.csv', index=False)
display(inv)
log(f'outputs_summary.txt written to {SUMMARY_PATH}')
log('Notebook run complete.')
print('Cell 12 is completed')
```
