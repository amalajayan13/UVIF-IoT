```python
# ── Cell 1 — Environment, paths, and run configuration ───────────────────────
print("Cell 1: setup entered")

import gc
import json
import random
import time
import warnings
from pathlib import Path

import numpy as np
import pandas as pd

warnings.filterwarnings("ignore")

SEED = 42
RUN_MODE = "final"          # "pilot" or "final"
PROJECT_NAME = "UVIF-IoT-WiMob2026"

random.seed(SEED)
np.random.seed(SEED)

try:
    import torch
    import torch.nn as nn
    import torch.nn.functional as F
    from torch.utils.data import TensorDataset, DataLoader

    torch.manual_seed(SEED)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(SEED)

    DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

except Exception as e:
    raise RuntimeError(f"PyTorch import failed: {e}")

try:
    from google.colab import drive
    IN_COLAB = True
except Exception:
    IN_COLAB = False


if IN_COLAB:
    try:
        drive.mount("/content/drive", force_remount=False)
        PROJECT_ROOT = Path("/content/drive/MyDrive") / PROJECT_NAME
    except Exception:
        PROJECT_ROOT = Path("/content") / PROJECT_NAME
else:
    PROJECT_ROOT = Path.cwd().resolve()


OUT_DIR      = PROJECT_ROOT / "outputs"
AUDIT_DIR    = OUT_DIR / "audit"
TABLE_DIR    = OUT_DIR / "tables"
FIG_DIR      = OUT_DIR / "figures"
MODEL_DIR    = OUT_DIR / "models"
CKPT_DIR     = OUT_DIR / "checkpoints"
FORENSIC_DIR = OUT_DIR / "forensics"
LOG_DIR      = OUT_DIR / "logs"

for d in [OUT_DIR, AUDIT_DIR, TABLE_DIR, FIG_DIR, MODEL_DIR, CKPT_DIR, FORENSIC_DIR, LOG_DIR]:
    d.mkdir(parents=True, exist_ok=True)


RUN_LOG_PATH = LOG_DIR / "run_log.txt"
SUMMARY_PATH = LOG_DIR / "run_summary.txt"
RUN_LOG_PATH.write_text("", encoding="utf-8")


def now():
    return time.strftime("%H:%M:%S")


def log(msg):
    line = f"[{now()}] {msg}"
    print(line)
    with open(RUN_LOG_PATH, "a", encoding="utf-8") as f:
        f.write(line + "\n")


def save_json(obj, path):
    with open(path, "w", encoding="utf-8") as f:
        json.dump(obj, f, indent=2)


CFG = {
    "max_total_rows": 300_000 if RUN_MODE == "pilot" else 1_500_000,
    "max_rows_per_file": 25_000 if RUN_MODE == "pilot" else 150_000,

    "batch_size": 512,
    "epochs": 30 if RUN_MODE == "pilot" else 50,
    "lr": 1e-3,
    "weight_decay": 1e-5,
    "dropout": 0.30,
    "early_stop": 8,

    "ece_bins": 15,
    "pgd_steps": 10,
    "pgd_alpha": 0.004,
    "adv_batch_cap": 50 if RUN_MODE == "pilot" else 200,

    "rf_trees": 100 if RUN_MODE == "pilot" else 300,
    "feature_budgets": [5, 10, 15, 20, 30, 50, 115],
}


DATA_ROOT_CANDIDATES = [
    PROJECT_ROOT / "data" / "nbaiot",
    PROJECT_ROOT / "data",
    PROJECT_ROOT.parent / "data" / "nbaiot",
    PROJECT_ROOT.parent / "data",
    PROJECT_ROOT,
]

if IN_COLAB:
    DATA_ROOT_CANDIDATES += [
        Path("/content/drive/MyDrive/Datasets/nbaiot"),
        Path("/content/drive/MyDrive/Datasets/N_BaIoT"),
        Path("/content/drive/MyDrive/Datasets"),
        Path("/content"),
    ]


setup = {
    "project": PROJECT_NAME,
    "run_mode": RUN_MODE,
    "seed": SEED,
    "in_colab": IN_COLAB,
    "project_root": str(PROJECT_ROOT),
    "device": str(DEVICE),
    "data_root_candidates": [str(p) for p in DATA_ROOT_CANDIDATES],
    "config": CFG,
}

save_json(setup, AUDIT_DIR / "setup.json")

log(f"Project root: {PROJECT_ROOT}")
log(f"Output root: {OUT_DIR}")
log(f"Run mode: {RUN_MODE}")
log(f"Device: {DEVICE}")

if DEVICE.type == "cuda":
    log(f"GPU: {torch.cuda.get_device_name(0)}")
else:
    log("GPU not detected. Training will be slower.")

log("Cell 1 completed")
```

```python
# ── Cell 2 — N-BaIoT discovery, loading, and audit ───────────────────────────
print("Cell 2: dataset loading entered")

def label_from_name(path):
    s = str(path).lower()

    if "benign" in s:
        return 0

    attacks = ["gafgyt", "mirai", "bashlite", "ack", "scan", "syn",
               "udp", "udpplain", "tcp", "junk", "combo", "http"]

    if any(a in s for a in attacks):
        return 1

    return None


def attack_from_name(path):
    s = str(path).lower()

    if "benign" in s:
        return "benign"

    attacks = ["gafgyt", "mirai", "bashlite", "ack", "scan", "syn",
               "udpplain", "udp", "tcp", "junk", "combo", "http"]

    found = [a for a in attacks if a in s]
    return "+".join(found) if found else "attack"


def device_from_name(path):
    stem = Path(path).stem
    first = stem.split(".")[0]

    if first.isdigit():
        return f"device_{first}"

    return "unknown"


def find_nbaiot_csvs(paths):
    files = []

    skip = {"outputs", ".git", ".ipynb_checkpoints", "__pycache__"}

    for root in paths:
        root = Path(root)

        if not root.exists():
            continue

        for fp in root.rglob("*.csv"):
            if any(part.lower() in skip for part in fp.parts):
                continue

            label = label_from_name(fp)

            if label is not None:
                files.append((fp.resolve(), label))

    return sorted(set(files), key=lambda x: str(x[0]).lower())


def read_sample(path, n):
    df = pd.read_csv(path, low_memory=False)

    if len(df) > n:
        df = df.sample(n=n, random_state=SEED)

    return df.reset_index(drop=True)


nb_files = find_nbaiot_csvs(DATA_ROOT_CANDIDATES)
log(f"N-BaIoT CSV files found: {len(nb_files)}")

if not nb_files:
    raise FileNotFoundError("No N-BaIoT CSV files found. Check DATA_ROOT_CANDIDATES in Cell 1.")


file_audit = pd.DataFrame([
    {
        "source_file": str(fp),
        "file_name": fp.name,
        "device": device_from_name(fp),
        "label": label,
        "label_name": "benign" if label == 0 else "attack",
        "attack_type": attack_from_name(fp),
        "size_mb": round(fp.stat().st_size / 1024**2, 4),
    }
    for fp, label in nb_files
])

file_audit.to_csv(AUDIT_DIR / "nbaiot_file_audit.csv", index=False)


frames = []
counts = {0: 0, 1: 0}
per_class_cap = CFG["max_total_rows"] // 2

for fp, label in nb_files:
    if counts[label] >= per_class_cap:
        continue

    remaining = per_class_cap - counts[label]
    n_take = min(CFG["max_rows_per_file"], remaining)

    if n_take <= 0:
        continue

    try:
        df = read_sample(fp, n_take)

        df["label"] = label
        df["label_name"] = "benign" if label == 0 else "attack"
        df["source_file"] = str(fp)
        df["source_file_name"] = fp.name
        df["device"] = device_from_name(fp)
        df["attack_type"] = attack_from_name(fp)

        frames.append(df)
        counts[label] += len(df)

        log(f"{fp.name:24s} {df['label_name'].iloc[0]:7s} rows={len(df):,} totals={counts}")

    except Exception as e:
        log(f"Skipped {fp}: {e}")

    if min(counts.values()) >= per_class_cap:
        break

    gc.collect()


if not frames:
    raise RuntimeError("CSV files were found, but none could be loaded.")


nbaiot_raw = pd.concat(frames, ignore_index=True)

min_count = nbaiot_raw["label"].value_counts().min()
nbaiot_raw = (
    nbaiot_raw
    .groupby("label", group_keys=False)
    .apply(lambda x: x.sample(n=min_count, random_state=SEED))
    .reset_index(drop=True)
)

meta_cols = ["label", "label_name", "source_file", "source_file_name", "device", "attack_type"]
feature_cols = [
    c for c in nbaiot_raw.columns
    if c not in meta_cols and pd.api.types.is_numeric_dtype(nbaiot_raw[c])
]

audit = {
    "rows": int(nbaiot_raw.shape[0]),
    "columns": int(nbaiot_raw.shape[1]),
    "features": int(len(feature_cols)),
    "label_counts": nbaiot_raw["label_name"].value_counts().to_dict(),
    "device_counts": nbaiot_raw["device"].value_counts().to_dict(),
    "attack_type_counts": nbaiot_raw["attack_type"].value_counts().to_dict(),
    "missing_values": int(nbaiot_raw[feature_cols].isna().sum().sum()),
    "duplicate_rows": int(nbaiot_raw.duplicated().sum()),
}

save_json(audit, AUDIT_DIR / "nbaiot_dataset_audit.json")
pd.DataFrame([audit]).to_csv(AUDIT_DIR / "nbaiot_dataset_audit.csv", index=False)

log(f"Loaded N-BaIoT shape: {nbaiot_raw.shape}")
log(f"Feature columns: {len(feature_cols)}")
log(f"Label counts: {audit['label_counts']}")
log(f"Missing feature values: {audit['missing_values']}")
log(f"Duplicate rows: {audit['duplicate_rows']}")
log("Cell 2 completed")

nbaiot_raw.head()
```

```python
# ── Cell 3 — Preprocessing, stratified split, and loaders ────────────────────
print("Cell 3: preprocessing entered")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

meta_cols = ["label", "label_name", "source_file", "source_file_name", "device", "attack_type"]

df = nbaiot_raw.replace([np.inf, -np.inf], np.nan).copy()
y = df["label"].astype(int).to_numpy()
idx = np.arange(len(df))

X_df = df[feature_cols].copy()

idx_train, idx_tmp, y_train, y_tmp = train_test_split(
    idx, y, test_size=0.30, stratify=y, random_state=SEED
)

idx_val, idx_test, y_val, y_test = train_test_split(
    idx_tmp, y_tmp, test_size=2/3, stratify=y_tmp, random_state=SEED
)

X_train_df = X_df.iloc[idx_train].copy()
X_val_df   = X_df.iloc[idx_val].copy()
X_test_df  = X_df.iloc[idx_test].copy()

valid_features = X_train_df.columns[X_train_df.notna().any(axis=0)].tolist()
medians = X_train_df[valid_features].median(numeric_only=True).fillna(0)

X_train_df = X_train_df[valid_features].fillna(medians)
X_val_df   = X_val_df[valid_features].fillna(medians)
X_test_df  = X_test_df[valid_features].fillna(medians)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train_df).astype(np.float32)
X_val   = scaler.transform(X_val_df).astype(np.float32)
X_test  = scaler.transform(X_test_df).astype(np.float32)

feature_cols = valid_features

meta_train = df.iloc[idx_train][meta_cols].reset_index(drop=False).rename(columns={"index": "original_index"})
meta_val   = df.iloc[idx_val][meta_cols].reset_index(drop=False).rename(columns={"index": "original_index"})
meta_test  = df.iloc[idx_test][meta_cols].reset_index(drop=False).rename(columns={"index": "original_index"})

meta_test.to_csv(AUDIT_DIR / "test_provenance.csv", index=False)
pd.DataFrame({"feature": feature_cols}).to_csv(AUDIT_DIR / "feature_columns.csv", index=False)


def make_loader(X, y, shuffle=False):
    ds = TensorDataset(
        torch.tensor(X, dtype=torch.float32),
        torch.tensor(y, dtype=torch.long)
    )
    return DataLoader(
        ds,
        batch_size=CFG["batch_size"],
        shuffle=shuffle,
        num_workers=0,
        pin_memory=(DEVICE.type == "cuda")
    )


train_loader = make_loader(X_train, y_train, True)
val_loader   = make_loader(X_val, y_val, False)
test_loader  = make_loader(X_test, y_test, False)

split_info = {
    "features": len(feature_cols),
    "train_shape": list(X_train.shape),
    "val_shape": list(X_val.shape),
    "test_shape": list(X_test.shape),
    "train_labels": {str(k): int(v) for k, v in zip(*np.unique(y_train, return_counts=True))},
    "val_labels": {str(k): int(v) for k, v in zip(*np.unique(y_val, return_counts=True))},
    "test_labels": {str(k): int(v) for k, v in zip(*np.unique(y_test, return_counts=True))},
}

save_json(split_info, AUDIT_DIR / "split_info.json")

log(f"Train: {X_train.shape}, Val: {X_val.shape}, Test: {X_test.shape}")
log(f"Train labels: {split_info['train_labels']}")
log(f"Val labels: {split_info['val_labels']}")
log(f"Test labels: {split_info['test_labels']}")
log("Cell 3 completed")
```

```python
# ── Cell 4 — Models and evaluation helpers ───────────────────────────────────
print("Cell 4: models and helpers entered")

import joblib

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    balanced_accuracy_score, roc_auc_score, average_precision_score,
    confusion_matrix, log_loss, brier_score_loss
)


class MLP(nn.Module):
    def __init__(self, in_dim, out_dim=2, hidden=(128, 64), dropout=0.2):
        super().__init__()

        layers = []
        prev = in_dim

        for h in hidden:
            layers += [
                nn.Linear(prev, h),
                nn.ReLU(),
                nn.Dropout(dropout),
            ]
            prev = h

        layers.append(nn.Linear(prev, out_dim))
        self.net = nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x)


class TemperatureScaler(nn.Module):
    def __init__(self):
        super().__init__()
        self.log_temp = nn.Parameter(torch.zeros(1))

    def forward(self, logits):
        temp = torch.exp(self.log_temp).clamp(0.05, 20.0)
        return logits / temp

    @property
    def temperature(self):
        return float(torch.exp(self.log_temp).detach().cpu().item())


def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)


def ece_score(y_true, probs, n_bins=15):
    conf = probs.max(axis=1)
    pred = probs.argmax(axis=1)
    correct = (pred == y_true).astype(float)

    bins = np.linspace(0, 1, n_bins + 1)
    ece = 0.0

    for lo, hi in zip(bins[:-1], bins[1:]):
        mask = (conf > lo) & (conf <= hi)

        if mask.sum() == 0:
            continue

        ece += mask.mean() * abs(correct[mask].mean() - conf[mask].mean())

    return float(ece)


def predict_torch(model, loader, scaler=None):
    model.eval()

    probs_all = []
    y_all = []

    with torch.no_grad():
        for x, y in loader:
            x = x.to(DEVICE)
            logits = model(x)

            if scaler is not None:
                logits = scaler(logits)

            probs = F.softmax(logits, dim=1)

            probs_all.append(probs.cpu().numpy())
            y_all.append(y.numpy())

    probs_all = np.vstack(probs_all)
    y_all = np.concatenate(y_all)

    return probs_all, probs_all.argmax(axis=1), y_all


def metric_row(y_true, y_pred, probs):
    out = {
        "accuracy": accuracy_score(y_true, y_pred),
        "balanced_accuracy": balanced_accuracy_score(y_true, y_pred),
        "precision": precision_score(y_true, y_pred, zero_division=0),
        "recall": recall_score(y_true, y_pred, zero_division=0),
        "f1": f1_score(y_true, y_pred, zero_division=0),
        "ece": ece_score(y_true, probs, CFG["ece_bins"]),
    }

    try:
        out["roc_auc"] = roc_auc_score(y_true, probs[:, 1])
    except Exception:
        out["roc_auc"] = np.nan

    try:
        out["pr_auc"] = average_precision_score(y_true, probs[:, 1])
    except Exception:
        out["pr_auc"] = np.nan

    try:
        out["nll"] = log_loss(y_true, probs, labels=[0, 1])
    except Exception:
        out["nll"] = np.nan

    try:
        out["brier"] = brier_score_loss(y_true, probs[:, 1])
    except Exception:
        out["brier"] = np.nan

    return out


def make_table_latex(df, path):
    try:
        df.to_latex(path, float_format="%.4f")
    except Exception as e:
        log(f"LaTeX export failed for {path.name}: {e}")


log("Model and metric helpers ready")
log("Cell 4 completed")
```

```python
# ── Cell 5 — Training: baselines and compact UVIF-IoT ────────────────────────
print("Cell 5: training entered")

if torch.cuda.is_available():
    torch.cuda.empty_cache()
gc.collect()


def train_mlp(model, train_loader, val_loader, epochs=None, tag="MLP"):
    epochs = epochs or CFG["epochs"]

    model.to(DEVICE)
    opt = torch.optim.Adam(
        model.parameters(),
        lr=CFG["lr"],
        weight_decay=CFG["weight_decay"]
    )

    best_f1 = -1
    best_state = None
    wait = 0
    history = []

    for ep in range(epochs):
        model.train()

        for x, y in train_loader:
            x = x.to(DEVICE)
            y = y.to(DEVICE)

            opt.zero_grad(set_to_none=True)
            loss = F.cross_entropy(model(x), y)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 5.0)
            opt.step()

        probs, preds, yt = predict_torch(model, val_loader)
        val_f1 = f1_score(yt, preds, zero_division=0)
        history.append(val_f1)

        if val_f1 > best_f1 + 1e-4:
            best_f1 = val_f1
            best_state = {k: v.detach().cpu().clone() for k, v in model.state_dict().items()}
            wait = 0
        else:
            wait += 1

        if ep == 0 or (ep + 1) % 5 == 0:
            log(f"[{tag}] epoch {ep + 1:02d}/{epochs} val_f1={val_f1:.4f}")

        if wait >= CFG["early_stop"]:
            log(f"[{tag}] early stop at epoch {ep + 1}; best_val_f1={best_f1:.4f}")
            break

    if best_state is not None:
        model.load_state_dict(best_state)

    return history


def subset_loader(X, y, cols, shuffle=False):
    return make_loader(X[:, cols], y, shuffle=shuffle)


MODELS = {}
HIST = {}

log("Training Logistic Regression")
lr = LogisticRegression(
    max_iter=1000,
    class_weight="balanced",
    n_jobs=-1,
    random_state=SEED
)
lr.fit(X_train, y_train)
MODELS["Logistic Regression"] = lr
joblib.dump(lr, MODEL_DIR / "logistic_regression.joblib")


log("Training Random Forest")
rf = RandomForestClassifier(
    n_estimators=CFG["rf_trees"],
    class_weight="balanced",
    random_state=SEED,
    n_jobs=-1,
    max_samples=0.7,
    bootstrap=True
)
rf.fit(X_train, y_train)
MODELS["Random Forest"] = rf
joblib.dump(rf, MODEL_DIR / "random_forest.joblib")


rf_rank = pd.DataFrame({
    "feature": feature_cols,
    "importance": rf.feature_importances_
}).sort_values("importance", ascending=False).reset_index(drop=True)

rf_rank["rank"] = np.arange(1, len(rf_rank) + 1)
rf_rank.to_csv(TABLE_DIR / "rf_feature_importance.csv", index=False)

top_k_uvif = 30
uvif_features = rf_rank.head(top_k_uvif)["feature"].tolist()
uvif_cols = [feature_cols.index(c) for c in uvif_features]

pd.DataFrame({"feature": uvif_features}).to_csv(
    TABLE_DIR / f"uvif_top_{top_k_uvif}_features.csv",
    index=False
)


log("Training Standard MLP")
mlp = MLP(
    in_dim=X_train.shape[1],
    out_dim=2,
    hidden=(128, 64),
    dropout=CFG["dropout"]
)

HIST["Standard MLP"] = train_mlp(
    mlp,
    train_loader,
    val_loader,
    tag="Standard MLP"
)

MODELS["Standard MLP"] = mlp
torch.save(
    {
        "state_dict": mlp.state_dict(),
        "features": feature_cols,
        "in_dim": X_train.shape[1],
    },
    CKPT_DIR / "standard_mlp.pt"
)


log(f"Training UVIF-IoT with top-{top_k_uvif} RF-ranked features")
uvif_train_loader = subset_loader(X_train, y_train, uvif_cols, shuffle=True)
uvif_val_loader   = subset_loader(X_val, y_val, uvif_cols, shuffle=False)
uvif_test_loader  = subset_loader(X_test, y_test, uvif_cols, shuffle=False)

uvif = MLP(
    in_dim=top_k_uvif,
    out_dim=2,
    hidden=(64, 32),
    dropout=CFG["dropout"]
)

HIST["UVIF-IoT"] = train_mlp(
    uvif,
    uvif_train_loader,
    uvif_val_loader,
    tag="UVIF-IoT"
)

MODELS["UVIF-IoT"] = uvif
torch.save(
    {
        "state_dict": uvif.state_dict(),
        "features": uvif_features,
        "feature_indices": uvif_cols,
        "in_dim": top_k_uvif,
    },
    CKPT_DIR / "uvif_iot_top30.pt"
)


hist_len = max(len(v) for v in HIST.values())
hist_df = pd.DataFrame({
    name: vals + [np.nan] * (hist_len - len(vals))
    for name, vals in HIST.items()
})
hist_df.index = np.arange(1, hist_len + 1)
hist_df.index.name = "epoch"
hist_df.to_csv(TABLE_DIR / "validation_f1_history.csv")

joblib.dump(scaler, MODEL_DIR / "standard_scaler.joblib")
save_json(feature_cols, MODEL_DIR / "feature_columns.json")
save_json(uvif_features, MODEL_DIR / "uvif_top30_features.json")

summary = {
    "models": list(MODELS.keys()),
    "full_features": len(feature_cols),
    "uvif_features": len(uvif_features),
    "feature_reduction_pct": 100 * (1 - len(uvif_features) / len(feature_cols)),
    "mlp_parameters": count_parameters(mlp),
    "uvif_parameters": count_parameters(uvif),
}

save_json(summary, AUDIT_DIR / "training_summary.json")

log(f"Models trained: {list(MODELS.keys())}")
log(f"UVIF-IoT feature reduction: {summary['feature_reduction_pct']:.1f}%")
log("Cell 5 completed")
```

```python
# ── Cell 6 — Clean and calibrated evaluation ─────────────────────────────────
print("Cell 6: clean and calibrated evaluation entered")

def predict_sklearn(model, X, y):
    probs = model.predict_proba(X)
    return probs, probs.argmax(axis=1), y


def fit_temperature(model, loader):
    model.eval()

    logits_all = []
    y_all = []

    with torch.no_grad():
        for x, y in loader:
            x = x.to(DEVICE)
            logits_all.append(model(x).detach())
            y_all.append(y.to(DEVICE))

    logits_all = torch.cat(logits_all)
    y_all = torch.cat(y_all)

    scaler_t = TemperatureScaler().to(DEVICE)
    opt = torch.optim.LBFGS([scaler_t.log_temp], lr=0.1, max_iter=50)

    def closure():
        opt.zero_grad()
        loss = F.cross_entropy(scaler_t(logits_all), y_all)
        loss.backward()
        return loss

    opt.step(closure)
    return scaler_t


RESULTS = {}
PROBS = {}
PREDS = {}
TARGETS = {}

# Logistic Regression
probs, preds, yt = predict_sklearn(MODELS["Logistic Regression"], X_test, y_test)
RESULTS["Logistic Regression"] = metric_row(yt, preds, probs)
PROBS["Logistic Regression"] = probs
PREDS["Logistic Regression"] = preds
TARGETS["Logistic Regression"] = yt

# Random Forest
probs, preds, yt = predict_sklearn(MODELS["Random Forest"], X_test, y_test)
RESULTS["Random Forest"] = metric_row(yt, preds, probs)
PROBS["Random Forest"] = probs
PREDS["Random Forest"] = preds
TARGETS["Random Forest"] = yt

# Standard MLP
probs, preds, yt = predict_torch(MODELS["Standard MLP"], test_loader)
RESULTS["Standard MLP"] = metric_row(yt, preds, probs)
PROBS["Standard MLP"] = probs
PREDS["Standard MLP"] = preds
TARGETS["Standard MLP"] = yt

mlp_temp = fit_temperature(MODELS["Standard MLP"], val_loader)
probs, preds, yt = predict_torch(MODELS["Standard MLP"], test_loader, scaler=mlp_temp)
RESULTS["Standard MLP + Temp"] = metric_row(yt, preds, probs)
PROBS["Standard MLP + Temp"] = probs
PREDS["Standard MLP + Temp"] = preds
TARGETS["Standard MLP + Temp"] = yt

# UVIF-IoT
probs, preds, yt = predict_torch(MODELS["UVIF-IoT"], uvif_test_loader)
RESULTS["UVIF-IoT"] = metric_row(yt, preds, probs)
PROBS["UVIF-IoT"] = probs
PREDS["UVIF-IoT"] = preds
TARGETS["UVIF-IoT"] = yt

uvif_temp = fit_temperature(MODELS["UVIF-IoT"], uvif_val_loader)
probs, preds, yt = predict_torch(MODELS["UVIF-IoT"], uvif_test_loader, scaler=uvif_temp)
RESULTS["UVIF-IoT + Temp"] = metric_row(yt, preds, probs)
PROBS["UVIF-IoT + Temp"] = probs
PREDS["UVIF-IoT + Temp"] = preds
TARGETS["UVIF-IoT + Temp"] = yt


clean_df = pd.DataFrame(RESULTS).T
clean_df.to_csv(TABLE_DIR / "clean_calibrated_metrics.csv")
make_table_latex(clean_df, TABLE_DIR / "table_clean_calibrated_metrics.tex")

for name in RESULTS:
    cm = confusion_matrix(TARGETS[name], PREDS[name], labels=[0, 1])
    cm_df = pd.DataFrame(
        cm,
        index=["true_benign", "true_attack"],
        columns=["pred_benign", "pred_attack"]
    )
    safe_name = name.lower().replace(" ", "_").replace("+", "plus").replace("-", "_")
    cm_df.to_csv(TABLE_DIR / f"confusion_matrix_{safe_name}.csv")

temperature_info = {
    "standard_mlp_temperature": mlp_temp.temperature,
    "uvif_iot_temperature": uvif_temp.temperature,
}

save_json(temperature_info, TABLE_DIR / "temperature_scaling.json")

for name, row in clean_df.iterrows():
    log(
        f"{name:22s} "
        f"Acc={row['accuracy']:.4f} "
        f"F1={row['f1']:.4f} "
        f"ROC-AUC={row['roc_auc']:.4f} "
        f"ECE={row['ece']:.4f}"
    )

log(f"Standard MLP temperature: {mlp_temp.temperature:.4f}")
log(f"UVIF-IoT temperature: {uvif_temp.temperature:.4f}")
log("Cell 6 completed")

clean_df
```

```python
# ── Cell 7 — Uncertainty scoring and forensic triage ledger ──────────────────
print("Cell 7: uncertainty and forensic triage entered")

def uncertainty_table(name):
    probs = PROBS[name]
    preds = PREDS[name]
    ytrue = TARGETS[name]

    confidence = probs.max(axis=1)
    attack_prob = probs[:, 1]
    entropy = -(probs * np.log(probs + 1e-12)).sum(axis=1)
    entropy_norm = entropy / np.log(probs.shape[1])
    uncertainty = 1.0 - confidence
    correct = (preds == ytrue).astype(int)

    out = meta_test.copy()
    out["true_label"] = ytrue
    out["pred_label"] = preds
    out["true_label_name"] = np.where(ytrue == 1, "attack", "benign")
    out["pred_label_name"] = np.where(preds == 1, "attack", "benign")
    out["attack_probability"] = attack_prob
    out["confidence"] = confidence
    out["confidence_uncertainty"] = uncertainty
    out["predictive_entropy"] = entropy
    out["predictive_entropy_norm"] = entropy_norm
    out["correct"] = correct

    out["forensic_priority_score"] = (
        0.50 * out["attack_probability"]
        + 0.30 * out["predictive_entropy_norm"]
        + 0.20 * out["confidence_uncertainty"]
    )

    out["model_name"] = name
    return out


ledger = uncertainty_table("UVIF-IoT + Temp")
ledger_path = FORENSIC_DIR / "suspicious_event_ledger.csv"
ledger.to_csv(ledger_path, index=False)

top_uncertain = ledger.sort_values(
    ["predictive_entropy_norm", "forensic_priority_score"],
    ascending=False
).head(100)

high_risk_uncertain = ledger[
    (ledger["pred_label"] == 1) & 
    (ledger["predictive_entropy_norm"] >= ledger["predictive_entropy_norm"].quantile(0.95))
].sort_values("forensic_priority_score", ascending=False).head(100)

false_negative = ledger[
    (ledger["true_label"] == 1) & 
    (ledger["pred_label"] == 0)
].sort_values("confidence", ascending=False).head(100)

false_positive = ledger[
    (ledger["true_label"] == 0) & 
    (ledger["pred_label"] == 1)
].sort_values("confidence", ascending=False).head(100)

top_priority = ledger.sort_values(
    "forensic_priority_score",
    ascending=False
).head(100)

top_uncertain.to_csv(FORENSIC_DIR / "top_uncertain_predictions.csv", index=False)
high_risk_uncertain.to_csv(FORENSIC_DIR / "high_risk_uncertain_events.csv", index=False)
false_negative.to_csv(FORENSIC_DIR / "high_confidence_false_negatives.csv", index=False)
false_positive.to_csv(FORENSIC_DIR / "high_confidence_false_positives.csv", index=False)
top_priority.to_csv(FORENSIC_DIR / "top_forensic_priority_events.csv", index=False)

summary_by_attack = (
    ledger
    .groupby(["device", "attack_type", "true_label_name"])
    .agg(
        samples=("true_label", "size"),
        detection_rate=("correct", "mean"),
        mean_attack_probability=("attack_probability", "mean"),
        mean_confidence=("confidence", "mean"),
        mean_entropy=("predictive_entropy_norm", "mean"),
        mean_priority=("forensic_priority_score", "mean"),
    )
    .reset_index()
)

summary_by_attack.to_csv(FORENSIC_DIR / "device_attack_triage_summary.csv", index=False)
make_table_latex(summary_by_attack.head(20), FORENSIC_DIR / "table_device_attack_triage_summary.tex")

unc_summary = {
    "ledger_rows": int(len(ledger)),
    "mean_confidence": float(ledger["confidence"].mean()),
    "mean_entropy_norm": float(ledger["predictive_entropy_norm"].mean()),
    "mean_priority": float(ledger["forensic_priority_score"].mean()),
    "false_negatives": int(((ledger["true_label"] == 1) & (ledger["pred_label"] == 0)).sum()),
    "false_positives": int(((ledger["true_label"] == 0) & (ledger["pred_label"] == 1)).sum()),
    "top_uncertain_saved": str(FORENSIC_DIR / "top_uncertain_predictions.csv"),
    "priority_formula": "0.50*attack_probability + 0.30*entropy_normalized + 0.20*confidence_uncertainty",
}

save_json(unc_summary, FORENSIC_DIR / "forensic_triage_summary.json")

log(f"Forensic ledger saved: {ledger_path}")
log(f"False negatives: {unc_summary['false_negatives']}")
log(f"False positives: {unc_summary['false_positives']}")
log(f"Mean normalized entropy: {unc_summary['mean_entropy_norm']:.6f}")
log("Cell 7 completed")

summary_by_attack.head(20)
```

```python
# ── Cell 8 — Adversarial robustness evaluation ──────────────────────────────
print("Cell 8: adversarial robustness entered")

def fgsm_attack(model, x, y, eps):
    x_adv = x.detach().clone()
    x_adv.requires_grad_(True)

    loss = F.cross_entropy(model(x_adv), y)
    grad = torch.autograd.grad(loss, x_adv)[0]

    return (x_adv + eps * grad.sign()).detach()


def pgd_attack(model, x, y, eps, alpha, steps):
    x0 = x.detach()
    x_adv = x0.clone()

    for _ in range(steps):
        x_adv.requires_grad_(True)

        loss = F.cross_entropy(model(x_adv), y)
        grad = torch.autograd.grad(loss, x_adv)[0]

        x_adv = x_adv.detach() + alpha * grad.sign()
        x_adv = torch.max(torch.min(x_adv, x0 + eps), x0 - eps)

    return x_adv.detach()


def adv_f1(model, loader, attack, eps, batch_cap=None):
    model.eval()

    preds = []
    targets = []

    for i, (x, y) in enumerate(loader):
        if batch_cap is not None and i >= batch_cap:
            break

        x = x.to(DEVICE)
        y = y.to(DEVICE)

        if attack == "fgsm":
            x_adv = fgsm_attack(model, x, y, eps)
        elif attack == "pgd":
            x_adv = pgd_attack(
                model,
                x,
                y,
                eps=eps,
                alpha=CFG["pgd_alpha"],
                steps=CFG["pgd_steps"]
            )
        else:
            raise ValueError("attack must be fgsm or pgd")

        with torch.no_grad():
            pred = model(x_adv).argmax(1)

        preds.extend(pred.cpu().numpy())
        targets.extend(y.cpu().numpy())

    return f1_score(targets, preds, zero_division=0)


def clean_subset_f1(model, loader, batch_cap=None):
    model.eval()

    preds = []
    targets = []

    with torch.no_grad():
        for i, (x, y) in enumerate(loader):
            if batch_cap is not None and i >= batch_cap:
                break

            x = x.to(DEVICE)
            pred = model(x).argmax(1)

            preds.extend(pred.cpu().numpy())
            targets.extend(y.numpy())

    return f1_score(targets, preds, zero_division=0)


adv_rows = []
eps_values = [0.005, 0.01, 0.02]
batch_cap = CFG["adv_batch_cap"]

adv_models = {
    "Standard MLP": (MODELS["Standard MLP"], test_loader),
    "UVIF-IoT": (MODELS["UVIF-IoT"], uvif_test_loader),
}

for name, (model, loader) in adv_models.items():
    model.to(DEVICE)

    row = {
        "model": name,
        "clean_subset_f1": clean_subset_f1(model, loader, batch_cap=batch_cap),
    }

    for eps in eps_values:
        row[f"fgsm_eps_{eps}"] = adv_f1(model, loader, "fgsm", eps, batch_cap=batch_cap)
        row[f"pgd_eps_{eps}"] = adv_f1(model, loader, "pgd", eps, batch_cap=batch_cap)

    adv_rows.append(row)

    log(
        f"{name:14s} clean={row['clean_subset_f1']:.4f} "
        f"FGSM@0.01={row['fgsm_eps_0.01']:.4f} "
        f"PGD@0.02={row['pgd_eps_0.02']:.4f}"
    )

adv_df = pd.DataFrame(adv_rows).set_index("model")
adv_df.to_csv(TABLE_DIR / "adversarial_robustness.csv")
make_table_latex(adv_df, TABLE_DIR / "table_adversarial_robustness.tex")

log("Cell 8 completed")
adv_df
```

```python
# ── Cell 9 — Edge efficiency: size, parameters, and inference latency ────────
print("Cell 9: edge efficiency entered")

def file_size_kb(path):
    path = Path(path)
    return round(path.stat().st_size / 1024, 2) if path.exists() else np.nan


def time_torch_model(model, X, repeats=5, batch_size=1024, device="cpu"):
    model = model.to(device)
    model.eval()

    X_tensor = torch.tensor(X, dtype=torch.float32)
    n = len(X_tensor)

    # warm-up
    with torch.no_grad():
        for i in range(0, min(n, batch_size * 2), batch_size):
            _ = model(X_tensor[i:i + batch_size].to(device))

    if device == "cuda":
        torch.cuda.synchronize()

    times = []

    with torch.no_grad():
        for _ in range(repeats):
            t0 = time.perf_counter()

            for i in range(0, n, batch_size):
                xb = X_tensor[i:i + batch_size].to(device)
                _ = model(xb)

            if device == "cuda":
                torch.cuda.synchronize()

            times.append(time.perf_counter() - t0)

    mean_time = float(np.mean(times))
    ms_per_sample = 1000 * mean_time / n
    throughput = n / mean_time

    return ms_per_sample, throughput


def time_sklearn_model(model, X, repeats=5):
    # warm-up
    _ = model.predict_proba(X[: min(len(X), 2048)])

    times = []

    for _ in range(repeats):
        t0 = time.perf_counter()
        _ = model.predict_proba(X)
        times.append(time.perf_counter() - t0)

    mean_time = float(np.mean(times))
    ms_per_sample = 1000 * mean_time / len(X)
    throughput = len(X) / mean_time

    return ms_per_sample, throughput


latency_n = min(20000, len(X_test))
X_latency_full = X_test[:latency_n]
X_latency_uvif = X_test[:latency_n, uvif_cols]

edge_rows = []

# Logistic Regression
lr_ms, lr_thr = time_sklearn_model(MODELS["Logistic Regression"], X_latency_full)
edge_rows.append({
    "model": "Logistic Regression",
    "input_features": len(feature_cols),
    "feature_reduction_pct": 0.0,
    "parameters": np.nan,
    "model_size_kb": file_size_kb(MODEL_DIR / "logistic_regression.joblib"),
    "cpu_ms_per_sample": lr_ms,
    "cpu_samples_per_sec": lr_thr,
    "gpu_ms_per_sample": np.nan,
    "gpu_samples_per_sec": np.nan,
})

# Random Forest
rf_ms, rf_thr = time_sklearn_model(MODELS["Random Forest"], X_latency_full)
edge_rows.append({
    "model": "Random Forest",
    "input_features": len(feature_cols),
    "feature_reduction_pct": 0.0,
    "parameters": np.nan,
    "model_size_kb": file_size_kb(MODEL_DIR / "random_forest.joblib"),
    "cpu_ms_per_sample": rf_ms,
    "cpu_samples_per_sec": rf_thr,
    "gpu_ms_per_sample": np.nan,
    "gpu_samples_per_sec": np.nan,
})

# Standard MLP
mlp_cpu_ms, mlp_cpu_thr = time_torch_model(
    MODELS["Standard MLP"],
    X_latency_full,
    device="cpu"
)

if torch.cuda.is_available():
    mlp_gpu_ms, mlp_gpu_thr = time_torch_model(
        MODELS["Standard MLP"],
        X_latency_full,
        device="cuda"
    )
else:
    mlp_gpu_ms, mlp_gpu_thr = np.nan, np.nan

edge_rows.append({
    "model": "Standard MLP",
    "input_features": len(feature_cols),
    "feature_reduction_pct": 0.0,
    "parameters": count_parameters(MODELS["Standard MLP"]),
    "model_size_kb": file_size_kb(CKPT_DIR / "standard_mlp.pt"),
    "cpu_ms_per_sample": mlp_cpu_ms,
    "cpu_samples_per_sec": mlp_cpu_thr,
    "gpu_ms_per_sample": mlp_gpu_ms,
    "gpu_samples_per_sec": mlp_gpu_thr,
})

# UVIF-IoT
uvif_cpu_ms, uvif_cpu_thr = time_torch_model(
    MODELS["UVIF-IoT"],
    X_latency_uvif,
    device="cpu"
)

if torch.cuda.is_available():
    uvif_gpu_ms, uvif_gpu_thr = time_torch_model(
        MODELS["UVIF-IoT"],
        X_latency_uvif,
        device="cuda"
    )
else:
    uvif_gpu_ms, uvif_gpu_thr = np.nan, np.nan

edge_rows.append({
    "model": "UVIF-IoT",
    "input_features": len(uvif_cols),
    "feature_reduction_pct": 100 * (1 - len(uvif_cols) / len(feature_cols)),
    "parameters": count_parameters(MODELS["UVIF-IoT"]),
    "model_size_kb": file_size_kb(CKPT_DIR / "uvif_iot_top30.pt"),
    "cpu_ms_per_sample": uvif_cpu_ms,
    "cpu_samples_per_sec": uvif_cpu_thr,
    "gpu_ms_per_sample": uvif_gpu_ms,
    "gpu_samples_per_sec": uvif_gpu_thr,
})

edge_df = pd.DataFrame(edge_rows)
edge_df.to_csv(TABLE_DIR / "edge_efficiency_summary.csv", index=False)

try:
    edge_df.to_latex(TABLE_DIR / "table_edge_efficiency_summary.tex", index=False, float_format="%.4f")
except Exception as e:
    log(f"Edge efficiency LaTeX export failed: {e}")

log(f"Latency measured on {latency_n:,} test samples")
log("Edge efficiency table saved")
log("Cell 9 completed")

edge_df
```

```python
# ── Cell 10 — Paper-ready figures ────────────────────────────────────────────
print("Cell 10: paper figures entered")

import matplotlib.pyplot as plt

def save_paper_fig(name):
    plt.savefig(FIG_DIR / f"{name}.pdf", dpi=300, bbox_inches="tight")
    plt.savefig(FIG_DIR / f"{name}.png", dpi=300, bbox_inches="tight")


paper_models = [
    "Logistic Regression",
    "Random Forest",
    "Standard MLP + Temp",
    "UVIF-IoT + Temp",
]

paper_labels = ["LR", "RF", "MLP", "UVIF-IoT"]

paper_metrics = clean_df.loc[paper_models].copy()
paper_metrics.index = paper_labels


# Figure 1: Clean detection and calibration as two panels
fig, axes = plt.subplots(1, 2, figsize=(8.0, 3.4))

axes[0].bar(paper_metrics.index, paper_metrics["f1"])
axes[0].set_ylabel("F1-score")
axes[0].set_title("Clean detection")
axes[0].set_ylim(0.998, 1.0001)
axes[0].grid(axis="y", alpha=0.25)

for i, v in enumerate(paper_metrics["f1"]):
    axes[0].text(i, v + 0.00003, f"{v:.4f}", ha="center", va="bottom", fontsize=8)

axes[1].bar(paper_metrics.index, paper_metrics["ece"])
axes[1].set_ylabel("Expected calibration error")
axes[1].set_title("Calibration")
axes[1].set_ylim(0, max(0.0006, paper_metrics["ece"].max() * 1.35))
axes[1].grid(axis="y", alpha=0.25)

for i, v in enumerate(paper_metrics["ece"]):
    axes[1].text(i, v + 0.000015, f"{v:.4f}", ha="center", va="bottom", fontsize=8)

plt.tight_layout()
save_paper_fig("fig_clean_detection_calibration")
plt.show()


# Figure 2: PGD robustness curve
robust_plot = adv_df.copy()
eps = np.array([0.0, 0.005, 0.01, 0.02])

plt.figure(figsize=(6.4, 3.8))

for model_name in ["Standard MLP", "UVIF-IoT"]:
    vals = [
        robust_plot.loc[model_name, "clean_subset_f1"],
        robust_plot.loc[model_name, "pgd_eps_0.005"],
        robust_plot.loc[model_name, "pgd_eps_0.01"],
        robust_plot.loc[model_name, "pgd_eps_0.02"],
    ]
    plt.plot(eps, vals, marker="o", linewidth=1.8, label=model_name)

plt.xlabel("PGD perturbation budget")
plt.ylabel("F1-score")
plt.title("Adversarial robustness under PGD")
plt.ylim(0.9990, 1.0000)
plt.grid(alpha=0.25)
plt.legend(fontsize=8)
plt.tight_layout()
save_paper_fig("fig_pgd_robustness")
plt.show()


# Figure 3: Edge efficiency as two panels
edge_plot = edge_df.set_index("model").loc[["Standard MLP", "UVIF-IoT"]]

fig, axes = plt.subplots(1, 2, figsize=(8.0, 3.4))

axes[0].bar(edge_plot.index, edge_plot["parameters"])
axes[0].set_ylabel("Trainable parameters")
axes[0].set_title("Model complexity")
axes[0].grid(axis="y", alpha=0.25)

for i, v in enumerate(edge_plot["parameters"]):
    axes[0].text(i, v + max(edge_plot["parameters"]) * 0.015, f"{int(v):,}", ha="center", va="bottom", fontsize=8)

axes[1].bar(edge_plot.index, edge_plot["model_size_kb"])
axes[1].set_ylabel("Model size (KB)")
axes[1].set_title("Serialized size")
axes[1].grid(axis="y", alpha=0.25)

for i, v in enumerate(edge_plot["model_size_kb"]):
    axes[1].text(i, v + max(edge_plot["model_size_kb"]) * 0.015, f"{v:.1f}", ha="center", va="bottom", fontsize=8)

fig.text(
    0.5,
    -0.02,
    f"UVIF-IoT uses {len(uvif_cols)} of {len(feature_cols)} features "
    f"({100 * (1 - len(uvif_cols) / len(feature_cols)):.1f}% reduction).",
    ha="center",
    fontsize=9
)

plt.tight_layout()
save_paper_fig("fig_edge_efficiency")
plt.show()


# Figure 4: CPU inference latency
latency_plot = edge_df.set_index("model").loc[
    ["Logistic Regression", "Random Forest", "Standard MLP", "UVIF-IoT"]
].copy()

latency_plot.index = ["LR", "RF", "MLP", "UVIF-IoT"]

plt.figure(figsize=(6.4, 3.8))
plt.bar(latency_plot.index, latency_plot["cpu_ms_per_sample"])
plt.ylabel("CPU latency (ms/sample)")
plt.title("Inference latency")
plt.grid(axis="y", alpha=0.25)

for i, v in enumerate(latency_plot["cpu_ms_per_sample"]):
    plt.text(i, v + latency_plot["cpu_ms_per_sample"].max() * 0.015, f"{v:.6f}", ha="center", va="bottom", fontsize=8)

plt.tight_layout()
save_paper_fig("fig_cpu_latency")
plt.show()


# Figure 5: RF top-20 feature importance
top20 = rf_rank.head(20).iloc[::-1]

plt.figure(figsize=(7.2, 5.2))
plt.barh(top20["feature"], top20["importance"])
plt.xlabel("Random Forest importance")
plt.ylabel("Feature")
plt.title("Top-20 RF-ranked N-BaIoT features")
plt.grid(axis="x", alpha=0.25)
plt.tight_layout()
save_paper_fig("fig_rf_top20_features")
plt.show()


# Figure index
fig_index = pd.DataFrame([
    {
        "figure": "fig_clean_detection_calibration",
        "purpose": "Clean F1 and calibration comparison",
        "pdf": str(FIG_DIR / "fig_clean_detection_calibration.pdf"),
        "png": str(FIG_DIR / "fig_clean_detection_calibration.png"),
    },
    {
        "figure": "fig_pgd_robustness",
        "purpose": "PGD robustness comparison",
        "pdf": str(FIG_DIR / "fig_pgd_robustness.pdf"),
        "png": str(FIG_DIR / "fig_pgd_robustness.png"),
    },
    {
        "figure": "fig_edge_efficiency",
        "purpose": "Parameter and model-size comparison",
        "pdf": str(FIG_DIR / "fig_edge_efficiency.pdf"),
        "png": str(FIG_DIR / "fig_edge_efficiency.png"),
    },
    {
        "figure": "fig_cpu_latency",
        "purpose": "CPU inference latency comparison",
        "pdf": str(FIG_DIR / "fig_cpu_latency.pdf"),
        "png": str(FIG_DIR / "fig_cpu_latency.png"),
    },
    {
        "figure": "fig_rf_top20_features",
        "purpose": "Feature-importance evidence",
        "pdf": str(FIG_DIR / "fig_rf_top20_features.pdf"),
        "png": str(FIG_DIR / "fig_rf_top20_features.png"),
    },
])

fig_index.to_csv(TABLE_DIR / "paper_figure_index.csv", index=False)

log("Improved paper figures saved as PDF and PNG")
log("Cell 10 completed")

fig_index
```

```python
# ── Cell 11 — Final summary, paper tables, and artifact inventory ────────────
print("Cell 11: final summary entered")

# Compact paper tables
paper_clean = clean_df.loc[
    ["Logistic Regression", "Random Forest", "Standard MLP + Temp", "UVIF-IoT + Temp"],
    ["accuracy", "precision", "recall", "f1", "roc_auc", "pr_auc", "ece", "brier"]
].copy()

paper_clean.index = ["LR", "RF", "MLP", "UVIF-IoT"]

paper_clean.to_csv(TABLE_DIR / "paper_table_clean_metrics.csv")
make_table_latex(paper_clean, TABLE_DIR / "paper_table_clean_metrics.tex")


paper_robust = adv_df[
    ["clean_subset_f1", "fgsm_eps_0.01", "pgd_eps_0.01", "pgd_eps_0.02"]
].copy()

paper_robust.to_csv(TABLE_DIR / "paper_table_robustness.csv")
make_table_latex(paper_robust, TABLE_DIR / "paper_table_robustness.tex")


paper_edge = edge_df.set_index("model").loc[
    ["Standard MLP", "UVIF-IoT"],
    [
        "input_features",
        "feature_reduction_pct",
        "parameters",
        "model_size_kb",
        "cpu_ms_per_sample",
        "cpu_samples_per_sec",
    ]
].copy()

paper_edge.to_csv(TABLE_DIR / "paper_table_edge_efficiency.csv")
make_table_latex(paper_edge, TABLE_DIR / "paper_table_edge_efficiency.tex")


forensic_brief = {
    "test_events": int(len(ledger)),
    "false_negatives": int(((ledger["true_label"] == 1) & (ledger["pred_label"] == 0)).sum()),
    "false_positives": int(((ledger["true_label"] == 0) & (ledger["pred_label"] == 1)).sum()),
    "mean_confidence": float(ledger["confidence"].mean()),
    "mean_entropy_norm": float(ledger["predictive_entropy_norm"].mean()),
    "mean_forensic_priority": float(ledger["forensic_priority_score"].mean()),
    "priority_formula": "0.50*attack_probability + 0.30*entropy_normalized + 0.20*confidence_uncertainty",
}

save_json(forensic_brief, FORENSIC_DIR / "paper_forensic_brief.json")

paper_forensic = pd.DataFrame([forensic_brief])
paper_forensic.to_csv(FORENSIC_DIR / "paper_forensic_brief.csv", index=False)

try:
    paper_forensic.to_latex(
        FORENSIC_DIR / "paper_table_forensic_brief.tex",
        index=False,
        float_format="%.4f"
    )
except Exception as e:
    log(f"Forensic LaTeX export failed: {e}")


# Run summary
summary_lines = []

summary_lines.append("UVIF-IoT pilot run summary")
summary_lines.append("=" * 60)
summary_lines.append("")
summary_lines.append(f"Dataset: N-BaIoT")
summary_lines.append(f"Run mode: {RUN_MODE}")
summary_lines.append(f"Rows loaded: {len(nbaiot_raw):,}")
summary_lines.append(f"Full feature count: {len(feature_cols)}")
summary_lines.append(f"UVIF-IoT feature count: {len(uvif_cols)}")
summary_lines.append(f"UVIF-IoT feature reduction: {100 * (1 - len(uvif_cols) / len(feature_cols)):.1f}%")
summary_lines.append("")
summary_lines.append("Important note:")
summary_lines.append("These are pilot-mode results for pipeline verification. Final manuscript claims should be based on a final-mode run.")
summary_lines.append("")
summary_lines.append("Clean detection and calibration:")
for name, row in paper_clean.iterrows():
    summary_lines.append(
        f"  {name:10s} F1={row['f1']:.4f}, ROC-AUC={row['roc_auc']:.4f}, "
        f"PR-AUC={row['pr_auc']:.4f}, ECE={row['ece']:.4f}"
    )

summary_lines.append("")
summary_lines.append("Adversarial robustness:")
for name, row in paper_robust.iterrows():
    summary_lines.append(
        f"  {name:12s} clean={row['clean_subset_f1']:.4f}, "
        f"FGSM@0.01={row['fgsm_eps_0.01']:.4f}, "
        f"PGD@0.02={row['pgd_eps_0.02']:.4f}"
    )

summary_lines.append("")
summary_lines.append("Edge efficiency:")
for name, row in paper_edge.iterrows():
    summary_lines.append(
        f"  {name:12s} features={int(row['input_features'])}, "
        f"params={int(row['parameters']):,}, "
        f"size={row['model_size_kb']:.1f} KB, "
        f"CPU latency={row['cpu_ms_per_sample']:.6f} ms/sample"
    )

summary_lines.append("")
summary_lines.append("Forensic triage:")
summary_lines.append(f"  Ledger rows: {forensic_brief['test_events']:,}")
summary_lines.append(f"  False negatives: {forensic_brief['false_negatives']}")
summary_lines.append(f"  False positives: {forensic_brief['false_positives']}")
summary_lines.append(f"  Priority formula: {forensic_brief['priority_formula']}")
summary_lines.append("")
summary_lines.append("Recommended paper framing:")
summary_lines.append("  UVIF-IoT should be framed as a compact, calibrated, robustness-evaluated,")
summary_lines.append("  and forensic-triage-supportive IDS for IoT edge security.")
summary_lines.append("  Do not claim legal forensic completeness; use 'cyberforensics-supportive' or")
summary_lines.append("  'forensic triage support for IoT security analysts'.")

SUMMARY_PATH.write_text("\n".join(summary_lines), encoding="utf-8")


# Artifact inventory
artifact_dirs = [AUDIT_DIR, TABLE_DIR, FIG_DIR, MODEL_DIR, CKPT_DIR, FORENSIC_DIR, LOG_DIR]
inventory_rows = []

for folder in artifact_dirs:
    if not folder.exists():
        continue

    for fp in sorted(folder.glob("*")):
        if fp.is_file():
            inventory_rows.append({
                "category": folder.name,
                "filename": fp.name,
                "size_kb": round(fp.stat().st_size / 1024, 2),
                "path": str(fp),
            })

artifact_inventory = pd.DataFrame(inventory_rows)
artifact_inventory.to_csv(OUT_DIR / "artifact_inventory.csv", index=False)

log(f"Paper clean metrics saved: {TABLE_DIR / 'paper_table_clean_metrics.csv'}")
log(f"Paper robustness table saved: {TABLE_DIR / 'paper_table_robustness.csv'}")
log(f"Paper edge table saved: {TABLE_DIR / 'paper_table_edge_efficiency.csv'}")
log(f"Forensic brief saved: {FORENSIC_DIR / 'paper_forensic_brief.json'}")
log(f"Run summary saved: {SUMMARY_PATH}")
log(f"Artifact inventory saved: {OUT_DIR / 'artifact_inventory.csv'}")
log("Cell 11 completed")

artifact_inventory
```
